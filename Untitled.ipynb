```python
import os
from yt_dlp import YoutubeDL

SAVE_PATH = 'D:/web/cnn_dnn_rnn_gan/dataset/Aman'
if not os.path.exists(SAVE_PATH): os.makedirs(SAVE_PATH)

# Gunakan query yang lebih umum agar banyak hasil
search_queries = [
    "doa sebelum belajar islam",
    "doa setelah belajar islam",
    "doa masuk wc islam",
    "doa keluar wc islam",
    "doa masuk masjid",
    "doa keluar masjid",
    "doa ketika bersin",
    "doa ketika mendengar orang bersin",
    "doa turun hujan",
    "doa ketika bercermin islam",
    "doa memakai pakaian",
    "surat pendek al quran anak"
]

ydl_opts = {
    # Kita ambil video paling ringan saja agar cepat
    'format': 'best[height<=360]', 
    'outtmpl': f'{SAVE_PATH}/%(id)s.%(ext)s',
    'max_filesize': 15 * 1024 * 1024,
    # Longgarkan filter durasi (maks 90 detik)
    'match_filter': lambda info: None if info.get('duration') and info.get('duration') < 300 else 'Terlalu lama',
    'ignoreerrors': True,
    'no_warnings': True,
    'quiet': False,
}

def mine_safe_data():
    with YoutubeDL(ydl_opts) as ydl:
        for query in search_queries:
            print(f"\n--- Mencoba lagi: {query} ---")
            # Ambil 30 video per query agar probabilitas berhasil tinggi
            ydl.download([f"ytsearch30:{query}"])

if __name__ == "__main__":
    mine_safe_data()
```
