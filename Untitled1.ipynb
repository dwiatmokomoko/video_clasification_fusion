```python
import os
import cv2
import librosa
import whisper
import easyocr
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models
from moviepy import VideoFileClip
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

# ==========================================
# 1. KONFIGURASI & PATH
# ==========================================
BASE_PATH = 'D:/web/cnn_dnn_rnn_gan/dataset'
SAVE_DATA_PATH = 'dataset_features.pkl'
IMG_SIZE = (128, 128)
MAX_TEXT_LEN = 100

# Model Global
print("Memuat Model Whisper & OCR...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'])

# ==========================================
# 2. INSIGHT DATASET (INFO JUMLAH FILE)
# ==========================================
def get_dataset_info():
    info = []
    total_files = 0
    print("\n--- STATISTIK DATASET ---")
    for folder in os.listdir(BASE_PATH):
        p = os.path.join(BASE_PATH, folder)
        if os.path.isdir(p):
            count = len([f for f in os.listdir(p) if f.endswith(('.mp4', '.mkv', '.avi'))])
            info.append({'Folder': folder, 'Jumlah Video': count})
            total_files += count
            print(f"Folder: {folder:15} | Total: {count} video")
    print(f"Total Keseluruhan: {total_files} video")
    print("-------------------------\n")
    return pd.DataFrame(info)

# ==========================================
# 3. FUNGSI PREPROCESSING & EKSTRAKSI
# ==========================================
def extract_and_save_features():
    # WAJIB: Hapus atau abaikan cache jika ingin memproses ulang data 'Aman' yang baru
    if os.path.exists(SAVE_DATA_PATH):
        print(f"Memuat cache fitur dari {SAVE_DATA_PATH}...")
        with open(SAVE_DATA_PATH, 'rb') as f:
            return pickle.load(f)

    data_list = []
    for folder_name in os.listdir(BASE_PATH):
        folder_path = os.path.join(BASE_PATH, folder_name)
        if not os.path.isdir(folder_path): continue
        
        # LABELING: 0 = AMAN, 1 = TOXIC
        label = 0 if folder_name.lower() == 'aman' else 1
        
        for video_file in os.listdir(folder_path):
            if video_file.endswith(('.mp4', '.mkv', '.avi')):
                v_path = os.path.join(folder_path, video_file)
                try:
                    # 1. VISUAL (CNN Input) - Ambil Frame
                    cap = cv2.VideoCapture(v_path)
                    cap.set(cv2.CAP_PROP_POS_FRAMES, int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) // 2)
                    ret, frame = cap.read()
                    if ret:
                        frame = cv2.resize(cv2.convertScaleAbs(frame), IMG_SIZE)
                        frame_norm = frame.astype(np.float32) / 255.0
                    else:
                        frame_norm = np.zeros((*IMG_SIZE, 3))
                    cap.release()

                    # 2. AUDIO (CNN Spectrogram)
                    audio_tmp = "temp_audio.wav"
                    with VideoFileClip(v_path) as video:
                        video.audio.write_audiofile(audio_tmp, logger=None)
                    
                    # Transkrip & OCR (RNN Input)
                    transcription = stt_model.transcribe(audio_tmp)['text']
                    ocr_res = ocr_reader.readtext(frame if ret else np.zeros(IMG_SIZE), detail=0)
                    full_text = transcription + " " + " ".join(ocr_res)

                    # Mel-Spectrogram
                    y, sr = librosa.load(audio_tmp)
                    spec = librosa.feature.melspectrogram(y=y, sr=sr)
                    spec_db = cv2.resize(librosa.power_to_db(spec), IMG_SIZE)
                    spec_db = np.expand_dims(spec_db, axis=-1)

                    data_list.append({
                        'img': frame_norm, 'spec': spec_db, 
                        'text': full_text, 'label': label
                    })
                    if os.path.exists(audio_tmp): os.remove(audio_tmp)
                    print(f"✅ Berhasil: {video_file}")

                except Exception as e:
                    print(f"❌ Gagal {video_file}: {e}")

    with open(SAVE_DATA_PATH, 'wb') as f:
        pickle.dump(data_list, f)
    return data_list

# ==========================================
# 4. SKEMA MODEL FUSI (CNN + RNN + DNN)
# ==========================================
def build_fusion_model():
    # Branch Visual (CNN)
    in_img = layers.Input(shape=(128, 128, 3))
    v = layers.Conv2D(32, (3,3), activation='relu')(in_img)
    v = layers.MaxPooling2D()(v)
    v = layers.GlobalAveragePooling2D()(v)

    # Branch Audio (CNN)
    in_spec = layers.Input(shape=(128, 128, 1))
    a = layers.Conv2D(32, (3,3), activation='relu')(in_spec)
    a = layers.MaxPooling2D()(a)
    a = layers.GlobalAveragePooling2D()(a)

    # Branch Text (RNN/LSTM)
    in_txt = layers.Input(shape=(MAX_TEXT_LEN,))
    t = layers.Embedding(5000, 64)(in_txt)
    t = layers.LSTM(64)(t)

    # Fusion (DNN)
    merged = layers.concatenate([v, a, t])
    d = layers.Dense(128, activation='relu')(merged)
    d = layers.Dropout(0.3)(d)
    out = layers.Dense(1, activation='sigmoid')(d)

    model = models.Model(inputs=[in_img, in_spec, in_txt], outputs=out)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ==========================================
# 5. EKSEKUSI UTAMA
# ==========================================
if __name__ == "__main__":
    get_dataset_info() # Tampilkan Info Folder
    raw_data = extract_and_save_features()
    df = pd.DataFrame(raw_data)

    if not df.empty:
        # Tokenizer
        tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=5000)
        tokenizer.fit_on_texts(df['text'])
        X_txt = tf.keras.preprocessing.sequence.pad_sequences(
            tokenizer.texts_to_sequences(df['text']), maxlen=MAX_TEXT_LEN
        )
        
        X_img = np.array(df['img'].tolist())
        X_spec = np.array(df['spec'].tolist())
        y = df['label'].values

        # Split & Train
        res = train_test_split(X_img, X_spec, X_txt, y, test_size=0.2)
        X_tr_img, X_ts_img, X_tr_spec, X_ts_spec, X_tr_txt, X_ts_txt, y_tr, y_ts = res

        model = build_fusion_model()
        model.fit([X_tr_img, X_tr_spec, X_tr_txt], y_tr, epochs=10, batch_size=4, 
                  validation_data=([X_ts_img, X_ts_spec, X_ts_txt], y_ts))
```

```python
# ==========================================
# 6. FUNGSI VISUALISASI (LENGKAP)
# ==========================================

def plot_training_history(history):
    """Menampilkan grafik Loss dan Accuracy untuk memantau Overfitting"""
    plt.figure(figsize=(12, 5))
    
    # Plot Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy', color='blue')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy', color='orange')
    plt.title('Model Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss', color='blue')
    plt.plot(history.history['val_loss'], label='Val Loss', color='orange')
    plt.title('Model Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    """Menampilkan Matrix untuk melihat salah tebak di kelas mana"""
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Aman', 'Kasar'], yticklabels=['Aman', 'Kasar'])
    plt.title('Confusion Matrix')
    plt.ylabel('Aktual')
    plt.xlabel('Prediksi')
    plt.show()

def plot_sample_features(df, index=0):
    """Menampilkan contoh input yang masuk ke model (CNN & RNN)"""
    plt.figure(figsize=(12, 4))
    
    # 1. Contoh Input Visual (CNN Thumbnail)
    plt.subplot(1, 3, 1)
    plt.imshow(df['img'][index])
    plt.title("Input CNN: Thumbnail")
    plt.axis('off')
    
    # 2. Contoh Input Audio (CNN Spectrogram)
    plt.subplot(1, 3, 2)
    plt.imshow(df['spec'][index].squeeze(), cmap='viridis')
    plt.title("Input CNN: Mel-Spectrogram")
    plt.axis('off')
    
    # 3. Contoh Input Teks (RNN Context)
    plt.subplot(1, 3, 3)
    text_snippet = df['text'][index][:50] + "..."
    plt.text(0.1, 0.5, f"Transkrip:\n{text_snippet}", fontsize=10, wrap=True)
    plt.title("Input RNN: Text")
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

def visualize_predictions(model, X_test, y_test, tokenizer, num_samples=5):
    """Menampilkan hasil prediksi AI pada data tes secara acak"""
    X_img, X_spec, X_txt = X_test
    indices = np.random.choice(len(y_test), num_samples, replace=False)
    
    plt.figure(figsize=(15, num_samples * 3))
    
    for i, idx in enumerate(indices):
        # Prediksi
        pred_prob = model.predict([X_img[idx:idx+1], X_spec[idx:idx+1], X_txt[idx:idx+1]])[0][0]
        pred_label = "KASAR/TOXIC" if pred_prob > 0.5 else "AMAN"
        true_label = "KASAR/TOXIC" if y_test[idx] == 1 else "AMAN"
        color = 'green' if (pred_prob > 0.5) == y_test[idx] else 'red'

        # Tampilan
        plt.subplot(num_samples, 2, i*2 + 1)
        plt.imshow(X_img[idx])
        plt.title(f"True: {true_label} | Pred: {pred_label}\nProb: {pred_prob:.2f}", color=color)
        plt.axis('off')

        plt.subplot(num_samples, 2, i*2 + 2)
        # Decode text sederhana (hanya simulasi karena X_txt sudah dipad)
        plt.text(0, 0.5, f"Analisis Teks:\n(Cek transkrip di dataframe)", fontsize=10)
        plt.axis('off')

    plt.tight_layout()
    plt.show()

# ==========================================
# 7. UPDATE EKSEKUSI UTAMA
# ==========================================
if __name__ == "__main__":
    get_dataset_info()
    raw_data = extract_and_save_features()
    df = pd.DataFrame(raw_data)

    if not df.empty:
        # Tampilkan contoh ekstraksi dulu
        print("\nMenampilkan contoh data hasil ekstraksi...")
        plot_sample_features(df, index=0)

        # Preprocessing Data... (sama seperti sebelumnya)
        # [Bagian Tokenizer & Train-Test Split di sini]
        
        # Training
        model = build_fusion_model()
        history = model.fit([X_tr_img, X_tr_spec, X_tr_txt], y_tr, 
                            validation_data=([X_ts_img, X_ts_spec, X_ts_txt], y_ts),
                            epochs=15, batch_size=4)

        # 1. Tampilkan Visualisasi Loss & Accuracy
        plot_training_history(history)

        # 2. Tampilkan Confusion Matrix
        y_pred = (model.predict([X_ts_img, X_ts_spec, X_ts_txt]) > 0.5).astype(int)
        plot_confusion_matrix(y_ts, y_pred)

        # 3. Tampilkan Plot Prediksi Video
        print("\nMenampilkan Sampel Prediksi Video...")
        visualize_predictions(model, [X_ts_img, X_ts_spec, X_ts_txt], y_ts, tokenizer)
```

```python
import os
import cv2
import librosa
import whisper
import easyocr
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras import layers, models
from moviepy import VideoFileClip
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ==========================================
# 1. KONFIGURASI & PARAMETER (REFERENSI ROBUST)
# ==========================================
BASE_PATH = 'D:/web/cnn_dnn_rnn_gan/dataset'
SAVE_DATA_PATH = 'dataset_features_v2.pkl'
IMG_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 2.0  # Durasi tetap untuk spectrogram

# Inisialisasi Model AI (Global)
print("Memuat Model Whisper & OCR...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'])

# ==========================================
# 2. AUDIO HELPERS (DARI REFERENSI ADVANCED)
# ==========================================
def pad_or_trim(y, sr, duration):
    target_length = int(sr * duration)
    if len(y) < target_length:
        y = np.pad(y, (0, target_length - len(y)), mode='constant')
    else:
        y = y[:target_length]
    return y

def safe_trim(y, top_db=25):
    try:
        y_trimmed, _ = librosa.effects.trim(y, top_db=top_db)
        return y_trimmed if len(y_trimmed) > 0 else y
    except:
        return y

# ==========================================
# 3. EKSTRAKSI FITUR (ROBUST VERSION)
# ==========================================
def extract_and_save_features():
    if os.path.exists(SAVE_DATA_PATH):
        print(f"Memuat data fitur dari {SAVE_DATA_PATH}...")
        with open(SAVE_DATA_PATH, 'rb') as f:
            return pickle.load(f)

    data_list = []
    print(f"Memulai ekstraksi dari {BASE_PATH}...")

    for folder_name in os.listdir(BASE_PATH):
        folder_path = os.path.join(BASE_PATH, folder_name)
        if not os.path.isdir(folder_path): continue
        
        # Labeling: Aman = 0, Selain itu = 1 (Toxic)
        label = 0 if folder_name.lower() == 'aman' else 1
        
        for video_file in os.listdir(folder_path):
            if video_file.endswith(('.mp4', '.mkv', '.avi')):
                v_path = os.path.join(folder_path, video_file)
                try:
                    # A. Visual: Ambil Frame Tengah
                    cap = cv2.VideoCapture(v_path)
                    cap.set(cv2.CAP_PROP_POS_FRAMES, int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) // 2)
                    ret, frame = cap.read()
                    if ret:
                        frame = cv2.resize(cv2.convertScaleAbs(frame), IMG_SIZE)
                        frame_norm = frame.astype(np.float32) / 255.0
                    else:
                        frame_norm = np.zeros((*IMG_SIZE, 3))
                    cap.release()

                    # B. Audio: Mel-Spectrogram (Robust Shape)
                    audio_tmp = "temp_audio.wav"
                    with VideoFileClip(v_path) as video:
                        if video.audio is not None:
                            video.audio.write_audiofile(audio_tmp, logger=None)
                            y, sr = librosa.load(audio_tmp, sr=SR)
                            y = pad_or_trim(safe_trim(y), SR, DURATION)
                            
                            # Transkrip Whisper
                            transcription = stt_model.transcribe(audio_tmp)['text']
                        else:
                            y = np.zeros(int(SR * DURATION))
                            transcription = ""

                    # Spectrogram untuk CNN Audio
                    spec = librosa.feature.melspectrogram(y=y, sr=SR, n_mels=128)
                    spec_db = cv2.resize(librosa.power_to_db(spec), IMG_SIZE)
                    spec_db = np.expand_dims(spec_db, axis=-1)

                    # C. Teks: OCR + Transkrip
                    ocr_res = ocr_reader.readtext(frame if ret else np.zeros(IMG_SIZE), detail=0)
                    full_text = transcription + " " + " ".join(ocr_res)

                    data_list.append({
                        'img': frame_norm,
                        'spec': spec_db,
                        'text': full_text,
                        'label': label
                    })
                    if os.path.exists(audio_tmp): os.remove(audio_tmp)
                    print(f"Berhasil: {video_file}")

                except Exception as e:
                    print(f"Gagal {video_file}: {e}")

    with open(SAVE_DATA_PATH, 'wb') as f:
        pickle.dump(data_list, f)
    return data_list

# ==========================================
# 4. MODEL ARCHITECTURE (OPTIMIZED)
# ==========================================
def build_fusion_model():
    # Input Layers
    in_img = layers.Input(shape=(128, 128, 3), name="Visual_CNN")
    in_spec = layers.Input(shape=(128, 128, 1), name="Audio_CNN")
    in_txt = layers.Input(shape=(MAX_TEXT_LEN,), name="Text_RNN")

    # Visual Branch (CNN)
    v = layers.Conv2D(32, (3,3), padding='same', activation='relu')(in_img)
    v = layers.BatchNormalization()(v)
    v = layers.MaxPooling2D((2,2))(v)
    v = layers.Dropout(0.3)(v)
    v = layers.Flatten()(v)

    # Audio Branch (CNN)
    a = layers.Conv2D(32, (3,3), padding='same', activation='relu')(in_spec)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D((2,2))(a)
    a = layers.Dropout(0.3)(a)
    a = layers.Flatten()(a)

    # Text Branch (RNN)
    t = layers.Embedding(5000, 64)(in_txt)
    t = layers.LSTM(64, dropout=0.2)(t)

    # Fusion (DNN)
    merged = layers.concatenate([v, a, t])
    d = layers.Dense(256, activation='relu')(merged)
    d = layers.BatchNormalization()(d)
    d = layers.Dropout(0.5)(d)
    out = layers.Dense(1, activation='sigmoid')(d)

    model = models.Model(inputs=[in_img, in_spec, in_txt], outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0001), 
                  loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ==========================================
# 5. EKSEKUSI UTAMA
# ==========================================
if __name__ == "__main__":
    raw_data = extract_and_save_features()
    df = pd.DataFrame(raw_data)

    if not df.empty:
        # A. Balancing Data (1:1 Ratio)
        df_aman = df[df['label'] == 0]
        df_kasar = df[df['label'] == 1]
        n_samples = min(len(df_aman), len(df_kasar))
        
        df_balanced = pd.concat([
            df_aman.sample(n_samples, random_state=42),
            df_kasar.sample(n_samples, random_state=42)
        ]).sample(frac=1).reset_index(drop=True)
        print(f"Data Seimbang: {len(df_balanced)} sampel total.")

        # B. Tokenisasi Teks
        tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=5000)
        tokenizer.fit_on_texts(df_balanced['text'])
        X_txt = tf.keras.preprocessing.sequence.pad_sequences(
            tokenizer.texts_to_sequences(df_balanced['text']), maxlen=MAX_TEXT_LEN
        )

        X_img = np.array(df_balanced['img'].tolist())
        X_spec = np.array(df_balanced['spec'].tolist())
        y = df_balanced['label'].values

        # C. Split Data
        splits = train_test_split(X_img, X_spec, X_txt, y, test_size=0.2, random_state=42)
        X_tr_img, X_ts_img, X_tr_spec, X_ts_spec, X_tr_txt, X_ts_txt, y_tr, y_ts = splits

        # D. Training
        model = build_fusion_model()
        callbacks = [
            EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
            ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3)
        ]

        print("\nMemulai Training Teroptimasi...")
        history = model.fit(
            [X_tr_img, X_tr_spec, X_tr_txt], y_tr,
            validation_data=([X_ts_img, X_ts_spec, X_ts_txt], y_ts),
            epochs=30, batch_size=8, callbacks=callbacks
        )

        # E. Visualisasi Akhir
        plt.figure(figsize=(12, 5))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['loss'], label='Train Loss')
        plt.plot(history.history['val_loss'], label='Val Loss')
        plt.title('Loss History')
        plt.legend()

        plt.subplot(1, 2, 2)
        y_pred = (model.predict([X_ts_img, X_ts_spec, X_ts_txt]) > 0.5).astype(int)
        cm = confusion_matrix(y_ts, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', 
                    xticklabels=['Aman', 'Kasar'], yticklabels=['Aman', 'Kasar'])
        plt.title('Confusion Matrix')
        plt.show()

        print(classification_report(y_ts, y_pred))
        model.save('model_fusion_final.h5')
```
