```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings
import tensorflow as tf
import librosa, whisper, easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight
import seaborn as sns

warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data.npz"
MODEL_FILE = "best_model.h5"
HISTORY_FILE = "history.npz"

IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 20

# ===============================
# LOAD MODELS
# ===============================
print("Loading Whisper and OCR models...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=False)

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE, 3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame / 255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = f"temp_{os.getpid()}.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE, 1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR * DURATION: 
            y = np.pad(y, (0, SR * DURATION - len(y)))
        y = y[:SR * DURATION] + 0.005 * np.random.randn(SR * DURATION)
        
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-9)
        
        text = ""
        try: 
            result = stt_model.transcribe(tmp)
            text = result['text']
        except: pass
        
        if os.path.exists(tmp): os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except Exception as e:
        if os.path.exists(tmp): os.remove(tmp)
        return np.zeros((*SPEC_SIZE, 1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame * 255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except: return ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c) % 5000 for c in text][:MAX_TEXT_LEN]
    seq += [0] * (MAX_TEXT_LEN - len(seq))
    return np.array(seq)

# ===============================
# LOAD / CACHE DATA (Updated for Nested Folders)
# ===============================
def load_or_extract_data():
    if os.path.exists(PROCESSED_DATA_FILE):
        print("Loading cached data...")
        data = np.load(PROCESSED_DATA_FILE)
        return data['X_img'], data['X_spec'], data['X_txt'], data['y']
    
    X_img, X_spec, X_txt, y = [], [], [], []
    # Mapping folder utama ke label
    categories = {'nonkasar': 0, 'kasar': 1}
    
    for cat_name, label in categories.items():
        cat_path = os.path.join(BASE_DATASET, cat_name)
        if not os.path.exists(cat_path): continue
        
        print(f"Processing category: {cat_name}...")
        # Walk melalui semua sub-folder
        for root, dirs, files in os.walk(cat_path):
            for f in files:
                if f.lower().endswith((".mp4", ".avi", ".mov")):
                    full_path = os.path.join(root, f)
                    try:
                        frame = extract_frame(full_path)
                        spec, text_audio = extract_audio(full_path)
                        text_ocr = extract_text_ocr(frame)
                        txt_seq = text_to_sequence(text_audio + " " + text_ocr)
                        
                        X_img.append(frame)
                        X_spec.append(spec)
                        X_txt.append(txt_seq)
                        y.append(label)
                        print(f" Success: {f}")
                    except Exception as e:
                        print(f" Error processing {f}: {e}")
                        continue
                        
    X_img, X_spec, X_txt, y = np.array(X_img), np.array(X_spec), np.array(X_txt), np.array(y)
    np.savez(PROCESSED_DATA_FILE, X_img=X_img, X_spec=X_spec, X_txt=X_txt, y=y)
    return X_img, X_spec, X_txt, y

# ===============================
# ADAPTIVE FUSION LAYER
# ===============================
from tensorflow.keras import backend as K
@tf.keras.utils.register_keras_serializable()
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.dense = layers.Dense(3, activation='softmax')
    def call(self, inputs):
        v, a, t = inputs
        combined = layers.concatenate([v, a, t])
        weights = self.dense(combined)
        wv = K.expand_dims(weights[:, 0], -1)
        wa = K.expand_dims(weights[:, 1], -1)
        wt = K.expand_dims(weights[:, 2], -1)
        return layers.concatenate([v * wv, a * wa, t * wt])

# ===============================
# BUILD MODEL
# ===============================
def build_model():
    img_in = layers.Input(shape=(*IMG_SIZE, 3), name="visual_input")
    spec_in = layers.Input(shape=(*SPEC_SIZE, 1), name="audio_input")
    txt_in = layers.Input(shape=(MAX_TEXT_LEN,), name="text_input")
    
    # Visual branch
    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    base.trainable = False
    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(v)
    v = layers.Dropout(0.3)(v)
    
    # Audio branch
    a = layers.Conv2D(32, 3, activation='relu', padding='same', kernel_regularizer=regularizers.l2(1e-4))(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64, 3, activation='relu', padding='same', kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((a.shape[1]*a.shape[2], a.shape[3]))(a)
    a = layers.Bidirectional(layers.GRU(64, dropout=0.3))(a)
    a = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(a)
    
    # Text branch
    t = layers.Embedding(5000, 128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64, dropout=0.3))(t)
    t = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(t)
    
    # Fusion
    fused = AdaptiveFusion()([v, a, t])
    
    # Classifier
    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    
    model = models.Model([img_in, spec_in, txt_in], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ===============================
# TRAINING & EVALUASI UTILS
# ===============================
def plot_training_history(history):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Val')
    plt.title('Model Accuracy'); plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Val')
    plt.title('Model Loss'); plt.legend()
    plt.show()

# ===============================
# MAIN
# ===============================
def main():
    # 1. Load Data
    X_img, X_spec, X_txt, y = load_or_extract_data()
    
    # 2. Split Data
    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_txt_tr, X_txt_val, y_tr, y_val = train_test_split(
        X_img, X_spec, X_txt, y, test_size=0.2, stratify=y, random_state=42
    )
    
    # 3. Hitung Class Weight untuk Balancing
    weights = class_weight.compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_tr),
        y=y_tr
    )
    class_weight_dict = dict(enumerate(weights))
    print(f"Computed Class Weights: {class_weight_dict}")
    
    # 4. Model Building / Loading
    if os.path.exists(MODEL_FILE):
        print("Loading existing model...")
        model = tf.keras.models.load_model(MODEL_FILE, custom_objects={'AdaptiveFusion': AdaptiveFusion})
        history_data = None
    else:
        print("Building and training new model...")
        model = build_model()
        
        # Callbacks
        callbacks = [
            tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=7, restore_best_weights=True),
            tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3),
            tf.keras.callbacks.ModelCheckpoint(MODEL_FILE, monitor='val_loss', save_best_only=True)
        ]
        
        history = model.fit(
            [X_img_tr, X_spec_tr, X_txt_tr], y_tr,
            validation_data=([X_img_val, X_spec_val, X_txt_val], y_val),
            epochs=EPOCHS,
            batch_size=BATCH_SIZE,
            class_weight=class_weight_dict, # PENERAPAN CLASS WEIGHT
            callbacks=callbacks
        )
        np.savez(HISTORY_FILE, history=history.history)
        history_data = history

    # 5. Evaluation
    preds = model.predict([X_img_val, X_spec_val, X_txt_val])
    preds_class = (preds > 0.5).astype(int)
    
    print("\n--- Classification Report ---")
    print(classification_report(y_val, preds_class, target_names=['Non-Kasar', 'Kasar']))
    
    if history_data:
        plot_training_history(history_data)

if __name__ == "__main__":
    main()
```

```python
import random
import os
import numpy as np
import tensorflow as tf

# ============================
# Fungsi untuk memprediksi dan menampilkan hasil
# ============================
def predict_file(file_path, model):
    # Ekstrak fitur menggunakan fungsi yang sudah didefinisikan sebelumnya
    frame = extract_frame(file_path)
    spec, text_audio = extract_audio(file_path)
    text_ocr = extract_text_ocr(frame)
    txt_seq = text_to_sequence(text_audio + " " + text_ocr)

    # Lakukan prediksi (tambahkan dimensi batch)
    pred = model.predict([
        np.expand_dims(frame, axis=0), 
        np.expand_dims(spec, axis=0), 
        np.expand_dims(txt_seq, axis=0)
    ], verbose=0) # verbose=0 agar tidak memenuhi console dengan progress bar
    
    # Klasifikasi hasil prediksi
    # Karena sigmoid, 0.5 ke atas = 1 (Kasar), di bawah 0.5 = 0 (Non-Kasar)
    confidence = pred[0][0]
    category = "Kasar" if confidence >= 0.5 else "Non-Kasar"
    
    # Hitung nilai probabilitas (jika < 0.5, tampilkan probabilitas ke arah Non-Kasar)
    display_accuracy = confidence if confidence >= 0.5 else (1 - confidence)
    
    return category, display_accuracy

# ============================
# Fungsi untuk Testing pada Folder Baru
# ============================
def run_prediction_test(model, dataset_path, num_samples=5):
    # Kategori utama sesuai struktur folder baru Anda
    main_categories = ['kasar', 'nonkasar']
    
    print("="*50)
    print("SISTEM PENGUJIAN MODEL MULTIMODAL")
    print("="*50)

    for main_cat in main_categories:
        main_path = os.path.join(dataset_path, main_cat)
        if not os.path.exists(main_path):
            print(f"Peringatan: Folder {main_cat} tidak ditemukan di {main_path}")
            continue
            
        print(f"\n>>> MENGUJI KATEGORI UTAMA: {main_cat.upper()}")
        
        # List semua sub-folder (contoh: Anjing, Babi, dll atau sub-folder non-kasar)
        sub_folders = [d for d in os.listdir(main_path) if os.path.isdir(os.path.join(main_path, d))]
        
        for sub in sub_folders:
            sub_path = os.path.join(main_path, sub)
            files = [f for f in os.listdir(sub_path) if f.lower().endswith((".mp4", ".avi", ".mov"))]
            
            if not files:
                continue
                
            # Ambil sampel acak
            sample_size = min(len(files), num_samples)
            random_files = random.sample(files, sample_size)
            
            print(f"\n--- Sub-Folder: {sub} ({len(files)} file) ---")
            
            for file_name in random_files:
                file_path = os.path.join(sub_path, file_name)
                category, accuracy = predict_file(file_path, model)
                
                # Format warna sederhana untuk console (opsional)
                # Kasar = Kategori Prediksi, Accuracy = Tingkat Kepercayaan Model
                print(f"File: {file_name}")
                print(f"  Hasil Prediksi: {category} ({accuracy*100:.2f}%)")

# ============================
# EKSEKUSI PREDIKSI
# ============================
if os.path.exists(MODEL_FILE):
    print("Memuat model dan menjalankan pengujian...")
    # Pastikan AdaptiveFusion sudah didefinisikan di atas sebelum load
    model = tf.keras.models.load_model(MODEL_FILE, custom_objects={'AdaptiveFusion': AdaptiveFusion})
    
    # Path dataset Anda
    DATASET_PATH = r"D:\web\cnn_dnn_rnn_gan\Dataset"
    
    # Jalankan pengujian (mengambil 3 sampel acak per sub-folder agar tidak terlalu lama)
    run_prediction_test(model, DATASET_PATH, num_samples=3)
else:
    print(f"Model '{MODEL_FILE}' tidak ditemukan. Selesaikan training terlebih dahulu.")
```
