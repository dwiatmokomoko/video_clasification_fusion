```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings, random
import tensorflow as tf
import librosa, whisper, easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers, backend as K
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight
import seaborn as sns

warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data_optimized.npz"
MODEL_FILE = "best_multimodal_model.h5"

IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 20

# ===============================
# LOAD MODELS
# ===============================
print("Initializing AI Engines (Whisper, EasyOCR)...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=True) # Set True jika ada NVIDIA GPU

# ===============================
# FEATURE EXTRACTION (MULTIMODAL)
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE, 3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame / 255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = f"temp_proc_{os.getpid()}.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE, 1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR * DURATION: 
            y = np.pad(y, (0, SR * DURATION - len(y)))
        y = y[:SR * DURATION]
        
        # Audio Augmentation (Noise Injection)
        y = y + 0.002 * np.random.randn(len(y))
        
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-9)
        
        text = ""
        try: 
            text = stt_model.transcribe(tmp)['text']
        except: pass
        
        if os.path.exists(tmp): os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except:
        if os.path.exists(tmp): os.remove(tmp)
        return np.zeros((*SPEC_SIZE, 1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame * 255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except: return ""

def text_to_sequence(text):
    # Membersihkan karakter non-alpha
    text = re.sub(r'[^a-zA-Z ]', '', text)
    # Tokenisasi karakter sederhana (ASCII mapping)
    seq = [ord(c) % 5000 for c in text][:MAX_TEXT_LEN]
    seq += [0] * (MAX_TEXT_LEN - len(seq))
    return np.array(seq)

# ===============================
# DATA LOADER
# ===============================
def load_dataset():
    if os.path.exists(PROCESSED_DATA_FILE):
        print("Loading cached data from disk...")
        d = np.load(PROCESSED_DATA_FILE)
        return d['X_img'], d['X_spec'], d['X_txt'], d['y']
    
    X_img, X_spec, X_txt, y = [], [], [], []
    categories = {'nonkasar': 0, 'kasar': 1}
    
    for cat_name, label in categories.items():
        cat_path = os.path.join(BASE_DATASET, cat_name)
        if not os.path.exists(cat_path): continue
        
        for root, dirs, files in os.walk(cat_path):
            # Abaikan folder checkpoint
            if ".ipynb_checkpoints" in root: continue
            
            for f in files:
                if f.lower().endswith((".mp4", ".avi")):
                    full_path = os.path.join(root, f)
                    try:
                        img = extract_frame(full_path)
                        spec, txt_a = extract_audio(full_path)
                        txt_o = extract_text_ocr(img)
                        txt_final = text_to_sequence(txt_a + " " + txt_o)
                        
                        X_img.append(img); X_spec.append(spec)
                        X_txt.append(txt_final); y.append(label)
                        print(f"[LOADED] {f}")
                    except: continue
                        
    X_i, X_s, X_t, Y = np.array(X_img), np.array(X_spec), np.array(X_txt), np.array(y)
    np.savez(PROCESSED_DATA_FILE, X_img=X_i, X_spec=X_s, X_txt=X_t, y=Y)
    return X_i, X_s, X_t, Y

# ===============================
# ADAPTIVE FUSION COMPONENT
# ===============================
@tf.keras.utils.register_keras_serializable()
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        # Sesuai flowchart: Pembobotan Dinamis
        self.attention_net = models.Sequential([
            layers.Dense(64, activation='relu'),
            layers.Dense(3, activation='softmax') # 3 bobot untuk V, A, T
        ])

    def call(self, inputs):
        v, a, t = inputs
        combined = layers.concatenate([v, a, t])
        weights = self.attention_net(combined)
        
        wv = K.expand_dims(weights[:, 0], -1)
        wa = K.expand_dims(weights[:, 1], -1)
        wt = K.expand_dims(weights[:, 2], -1)
        
        return layers.concatenate([v * wv, a * wa, t * wt])

# ===============================
# MODEL ARCHITECTURE (DNN BASED)
# ===============================
def build_optimized_model():
    # 1. Visual Branch (CNN - EfficientNet)
    in_img = layers.Input(shape=(*IMG_SIZE, 3))
    base_cnn = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=in_img)
    base_cnn.trainable = False
    v = layers.GlobalAveragePooling2D()(base_cnn.output)
    v = layers.Dense(128, activation='relu')(v)

    # 2. Audio Branch (CNN + RNN)
    in_spec = layers.Input(shape=(*SPEC_SIZE, 1))
    a = layers.Conv2D(32, 3, activation='relu', padding='same')(in_spec)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64, 3, activation='relu', padding='same')(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Flatten()(a)
    a = layers.Dense(128, activation='relu')(a)

    # 3. Text Branch (Embedding + LSTM)
    in_txt = layers.Input(shape=(MAX_TEXT_LEN,))
    t = layers.Embedding(5000, 64)(in_txt)
    t = layers.Bidirectional(layers.LSTM(64))(t)
    t = layers.Dense(128, activation='relu')(t)

    # 4. Adaptive Multimodal Fusion
    fused = AdaptiveFusion()([v, a, t])

    # 5. Classification Layer (DNN Based)
    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01))(fused)
    x = layers.Dropout(0.5)(x)
    x = layers.BatchNormalization()(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model([in_img, in_spec, in_txt], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4), 
                  loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ===============================
# PREDICTION & TESTING FUNCTIONS
# ===============================
def predict_single(file_path, model):
    img = extract_frame(file_path)
    spec, txt_a = extract_audio(file_path)
    txt_o = extract_text_ocr(img)
    txt_seq = text_to_sequence(txt_a + " " + txt_o)
    
    pred = model.predict([
        np.expand_dims(img, 0), 
        np.expand_dims(spec, 0), 
        np.expand_dims(txt_seq, 0)
    ], verbose=0)[0][0]
    
    label = "Kasar" if pred >= 0.5 else "Non-Kasar"
    score = pred if pred >= 0.5 else (1-pred)
    return label, score

def run_mass_test(model):
    print("\n" + "="*60)
    print("RUNNING MASSIVE TEST (10 FILES PER SUB-FOLDER)")
    print("="*60)
    
    for main_cat in ['kasar', 'nonkasar']:
        main_path = os.path.join(BASE_DATASET, main_cat)
        if not os.path.exists(main_path): continue
        
        print(f"\n>>> CATEGORY: {main_cat.upper()}")
        
        # Dapatkan semua subfolder
        subs = [d for d in os.listdir(main_path) if os.path.isdir(os.path.join(main_path, d))]
        for sub in subs:
            if sub == ".ipynb_checkpoints": continue
            sub_full = os.path.join(main_path, sub)
            files = [f for f in os.listdir(sub_full) if f.lower().endswith(".mp4")]
            
            # Ambil maksimal 10 file
            test_files = random.sample(files, min(len(files), 10))
            print(f"\n--- Sub-Folder: {sub} ---")
            
            for fname in test_files:
                fpath = os.path.join(sub_full, fname)
                res, acc = predict_single(fpath, model)
                print(f"[{res}] {acc*100:.2f}% | File: {fname[:50]}...")

# ===============================
# MAIN EXECUTION
# ===============================
def main():
    X_i, X_s, X_t, Y = load_dataset()
    
    # Split & Weighting
    X_i_tr, X_i_val, X_s_tr, X_s_val, X_t_tr, X_t_val, Y_tr, Y_val = train_test_split(
        X_i, X_s, X_t, Y, test_size=0.2, stratify=Y, random_state=42)
    
    cw = class_weight.compute_class_weight('balanced', classes=np.unique(Y_tr), y=Y_tr)
    cw_dict = dict(enumerate(cw))

    if not os.path.exists(MODEL_FILE):
        print("Training Model...")
        model = build_optimized_model()
        model.fit(
            [X_i_tr, X_s_tr, X_t_tr], Y_tr,
            validation_data=([X_i_val, X_s_val, X_t_val], Y_val),
            epochs=EPOCHS, batch_size=BATCH_SIZE,
            class_weight=cw_dict,
            callbacks=[tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)]
        )
        model.save(MODEL_FILE)
    else:
        print("Loading existing best model...")
        model = tf.keras.models.load_model(MODEL_FILE, custom_objects={'AdaptiveFusion': AdaptiveFusion})

    # Jalankan Test 10 File per folder
    run_mass_test(model)

if __name__ == "__main__":
    main()
```
