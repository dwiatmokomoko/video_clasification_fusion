```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings, random
import tensorflow as tf
import librosa, whisper, easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers, backend as K
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight
import seaborn as sns

warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data_optimized.npz"
MODEL_FILE = "best_multimodal_model.h5"
HISTORY_FILE = "training_history.npz"

IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 20

# ===============================
# LOAD MODELS
# ===============================
print("Initializing AI Engines (Whisper, EasyOCR)...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=True)

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE, 3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame / 255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = f"temp_proc_{os.getpid()}.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE, 1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR * DURATION: y = np.pad(y, (0, SR * DURATION - len(y)))
        y = y[:SR * DURATION] + 0.002 * np.random.randn(SR * DURATION)
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-9)
        text = ""
        try: text = stt_model.transcribe(tmp)['text']
        except: pass
        if os.path.exists(tmp): os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except:
        if os.path.exists(tmp): os.remove(tmp)
        return np.zeros((*SPEC_SIZE, 1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame * 255).astype(np.uint8)
        return " ".join(ocr_reader.readtext(frame_uint8, detail=0)).lower()
    except: return ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c) % 5000 for c in text][:MAX_TEXT_LEN]
    seq += [0] * (MAX_TEXT_LEN - len(seq))
    return np.array(seq)

# ===============================
# DATA LOADER
# ===============================
def load_dataset():
    if os.path.exists(PROCESSED_DATA_FILE):
        d = np.load(PROCESSED_DATA_FILE)
        return d['X_img'], d['X_spec'], d['X_txt'], d['y']
    
    X_img, X_spec, X_txt, y = [], [], [], []
    categories = {'nonkasar': 0, 'kasar': 1}
    for cat_name, label in categories.items():
        cat_path = os.path.join(BASE_DATASET, cat_name)
        if not os.path.exists(cat_path): continue
        for root, dirs, files in os.walk(cat_path):
            if ".ipynb_checkpoints" in root: continue
            for f in files:
                if f.lower().endswith((".mp4", ".avi")):
                    full_path = os.path.join(root, f)
                    try:
                        img = extract_frame(full_path)
                        spec, txt_a = extract_audio(full_path)
                        txt_o = extract_text_ocr(img)
                        X_img.append(img); X_spec.append(spec)
                        X_txt.append(text_to_sequence(txt_a + " " + txt_o)); y.append(label)
                    except: continue
    X_i, X_s, X_t, Y = np.array(X_img), np.array(X_spec), np.array(X_txt), np.array(y)
    np.savez(PROCESSED_DATA_FILE, X_img=X_i, X_spec=X_s, X_txt=X_t, y=Y)
    return X_i, X_s, X_t, Y

# ===============================
# ADAPTIVE FUSION LAYER
# ===============================
@tf.keras.utils.register_keras_serializable()
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.attention_net = models.Sequential([
            layers.Dense(64, activation='relu'),
            layers.Dense(3, activation='softmax')
        ])
    def call(self, inputs):
        v, a, t = inputs
        combined = layers.concatenate([v, a, t])
        weights = self.attention_net(combined)
        wv, wa, wt = weights[:, 0:1], weights[:, 1:2], weights[:, 2:3]
        return layers.concatenate([v * wv, a * wa, t * wt])

# ===============================
# MODEL ARCHITECTURE
# ===============================
def build_optimized_model():
    in_img = layers.Input(shape=(*IMG_SIZE, 3))
    base_cnn = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=in_img)
    base_cnn.trainable = False
    v = layers.GlobalAveragePooling2D()(base_cnn.output)
    v = layers.Dense(128, activation='relu')(v)

    in_spec = layers.Input(shape=(*SPEC_SIZE, 1))
    a = layers.Conv2D(32, 3, activation='relu', padding='same')(in_spec)
    a = layers.MaxPooling2D()(a)
    a = layers.Flatten()(a)
    a = layers.Dense(128, activation='relu')(a)

    in_txt = layers.Input(shape=(MAX_TEXT_LEN,))
    t = layers.Embedding(5000, 64)(in_txt)
    t = layers.Bidirectional(layers.LSTM(64))(t)
    t = layers.Dense(128, activation='relu')(t)

    fused = AdaptiveFusion()([v, a, t])
    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.01))(fused)
    x = layers.Dropout(0.5)(x)
    x = layers.BatchNormalization()(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model([in_img, in_spec, in_txt], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4), loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ===============================
# VISUALIZATION FUNCTIONS
# ===============================
def plot_history(history_dict):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history_dict['accuracy'], label='Train Acc')
    plt.plot(history_dict['val_accuracy'], label='Val Acc')
    plt.title('Training & Validation Accuracy'); plt.legend()
    
    plt.subplot(1, 2, 2)
    plt.plot(history_dict['loss'], label='Train Loss')
    plt.plot(history_dict['val_loss'], label='Val Loss')
    plt.title('Training & Validation Loss'); plt.legend()
    plt.show()

def plot_cm(y_true, y_pred):
    cm = confusion_matrix(y_true, (y_pred > 0.5).astype(int))
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Non-Kasar', 'Kasar'], yticklabels=['Non-Kasar', 'Kasar'])
    plt.xlabel('Predicted Label'); plt.ylabel('True Label')
    plt.title('Confusion Matrix Final')
    plt.show()

# ===============================
# MAIN EXECUTION
# ===============================
def main():
    X_i, X_s, X_t, Y = load_dataset()
    X_i_tr, X_i_val, X_s_tr, X_s_val, X_t_tr, X_t_val, Y_tr, Y_val = train_test_split(
        X_i, X_s, X_t, Y, test_size=0.2, stratify=Y, random_state=42)
    
    if not os.path.exists(MODEL_FILE):
        print("Training Model...")
        cw = dict(enumerate(class_weight.compute_class_weight('balanced', classes=np.unique(Y_tr), y=Y_tr)))
        model = build_optimized_model()
        history = model.fit([X_i_tr, X_s_tr, X_t_tr], Y_tr, 
                            validation_data=([X_i_val, X_s_val, X_t_val], Y_val),
                            epochs=EPOCHS, batch_size=BATCH_SIZE, class_weight=cw,
                            callbacks=[tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)])
        model.save(MODEL_FILE)
        plot_history(history.history)
    else:
        print("Loading model and visualizing...")
        model = tf.keras.models.load_model(MODEL_FILE, custom_objects={'AdaptiveFusion': AdaptiveFusion})

    # Evaluasi & CM
    preds = model.predict([X_i_val, X_s_val, X_t_val])
    print(classification_report(Y_val, (preds > 0.5).astype(int)))
    plot_cm(Y_val, preds)

    # TEST 10 FILE PER FOLDER
    print("\n>>> MENGUJI 10 FILE PER SUB-FOLDER")
    for main_cat in ['kasar', 'nonkasar']:
        main_path = os.path.join(BASE_DATASET, main_cat)
        for sub in [d for d in os.listdir(main_path) if os.path.isdir(os.path.join(main_path, d))]:
            if sub == ".ipynb_checkpoints": continue
            sub_full = os.path.join(main_path, sub)
            files = [f for f in os.listdir(sub_full) if f.lower().endswith(".mp4")]
            test_files = random.sample(files, min(len(files), 10))
            print(f"\n--- Folder: {sub} ---")
            for fname in test_files:
                img = extract_frame(os.path.join(sub_full, fname))
                spec, txt_a = extract_audio(os.path.join(sub_full, fname))
                p = model.predict([np.expand_dims(img,0), np.expand_dims(spec,0), np.expand_dims(text_to_sequence(txt_a),0)], verbose=0)[0][0]
                print(f"[{'Kasar' if p>=0.5 else 'Non-Kasar'}] {p*100:.2f}% | {fname[:40]}...")

if __name__ == "__main__":
    main()
```

```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings, random
import tensorflow as tf
import librosa, whisper, easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers, backend as K
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import class_weight
import seaborn as sns

warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data_optimized.npz"
MODEL_FILE = "best_multimodal_model.h5"
HISTORY_FILE = "training_history.npz" # File untuk simpan history

IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 20

# ===============================
# AI ENGINES
# ===============================
print("Initializing AI Engines...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=True)

# ===============================
# FUNCTIONS
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE, 3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame / 255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = f"temp_{os.getpid()}.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE, 1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR * DURATION: y = np.pad(y, (0, SR * DURATION - len(y)))
        y = y[:SR * DURATION]
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-9)
        text = stt_model.transcribe(tmp, fp16=False)['text']
        if os.path.exists(tmp): os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except:
        if os.path.exists(tmp): os.remove(tmp)
        return np.zeros((*SPEC_SIZE, 1)), ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c) % 5000 for c in text][:MAX_TEXT_LEN]
    seq += [0] * (MAX_TEXT_LEN - len(seq))
    return np.array(seq)

@tf.keras.utils.register_keras_serializable()
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.attention_net = models.Sequential([
            layers.Dense(64, activation='relu'),
            layers.Dense(3, activation='softmax')
        ])
    def call(self, inputs):
        v, a, t = inputs
        combined = layers.concatenate([v, a, t])
        weights = self.attention_net(combined)
        return layers.concatenate([v * weights[:, 0:1], a * weights[:, 1:2], t * weights[:, 2:3]])

def build_model():
    in_i = layers.Input(shape=(*IMG_SIZE, 3))
    v = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=in_i)
    v.trainable = False
    v = layers.GlobalAveragePooling2D()(v.output)
    v = layers.Dense(128, activation='relu')(v)

    in_s = layers.Input(shape=(*SPEC_SIZE, 1))
    a = layers.Conv2D(32, 3, activation='relu')(in_s)
    a = layers.Flatten()(a)
    a = layers.Dense(128, activation='relu')(a)

    in_t = layers.Input(shape=(MAX_TEXT_LEN,))
    t = layers.Embedding(5000, 64)(in_t)
    t = layers.Bidirectional(layers.LSTM(64))(t)
    t = layers.Dense(128, activation='relu')(t)

    fused = AdaptiveFusion()([v, a, t])
    x = layers.Dense(256, activation='relu')(fused)
    x = layers.Dropout(0.5)(x)
    out = layers.Dense(1, activation='sigmoid')(x)
    
    model = models.Model([in_i, in_s, in_t], out)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ===============================
# VISUALIZATION LOGIC
# ===============================
def show_plots(history_data, y_true=None, y_pred=None):
    # 1. Plot Accuracy & Loss
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    ax1.plot(history_data['accuracy'], label='Train Accuracy')
    ax1.plot(history_data['val_accuracy'], label='Val Accuracy')
    ax1.set_title('Model Accuracy')
    ax1.legend()

    ax2.plot(history_data['loss'], label='Train Loss')
    ax2.plot(history_data['val_loss'], label='Val Loss')
    ax2.set_title('Model Loss')
    ax2.legend()
    plt.show()

    # 2. Plot Confusion Matrix
    if y_true is not None and y_pred is not None:
        cm = confusion_matrix(y_true, (y_pred > 0.5).astype(int))
        plt.figure(figsize=(6, 5))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                    xticklabels=['Non-Kasar', 'Kasar'], yticklabels=['Non-Kasar', 'Kasar'])
        plt.title('Confusion Matrix Final')
        plt.show()

# ===============================
# MAIN
# ===============================
def main():
    # Load Data
    if not os.path.exists(PROCESSED_DATA_FILE):
        print("Data belum di-preprocess. Silakan jalankan ekstraksi data dahulu.")
        return
    
    data = np.load(PROCESSED_DATA_FILE)
    Xi, Xs, Xt, Y = data['X_img'], data['X_spec'], data['X_txt'], data['y']
    Xi_tr, Xi_val, Xs_tr, Xs_val, Xt_tr, Xt_val, Y_tr, Y_val = train_test_split(Xi, Xs, Xt, Y, test_size=0.2, stratify=Y)

    # Training atau Load Model
    if not os.path.exists(MODEL_FILE):
        print("Training model baru...")
        model = build_model()
        history = model.fit([Xi_tr, Xs_tr, Xt_tr], Y_tr, 
                            validation_data=([Xi_val, Xs_val, Xt_val], Y_val), 
                            epochs=EPOCHS, batch_size=BATCH_SIZE)
        model.save(MODEL_FILE)
        # Simpan history ke disk
        np.savez(HISTORY_FILE, **history.history)
        h_data = history.history
    else:
        print("Memuat model dan history yang sudah ada...")
        model = tf.keras.models.load_model(MODEL_FILE, custom_objects={'AdaptiveFusion': AdaptiveFusion})
        h_data = dict(np.load(HISTORY_FILE))

    # Tampilkan Grafik
    val_preds = model.predict([Xi_val, Xs_val, Xt_val])
    show_plots(h_data, Y_val, val_preds)

    # TEST 10 FILE PER FOLDER
    print("\n>>> TESTING 10 FILES PER SUBFOLDER")
    for cat in ['kasar', 'nonkasar']:
        p = os.path.join(BASE_DATASET, cat)
        for sub in [d for d in os.listdir(p) if os.path.isdir(os.path.join(p, d))]:
            if sub == ".ipynb_checkpoints": continue
            files = [f for f in os.listdir(os.path.join(p, sub)) if f.endswith(".mp4")]
            for f in random.sample(files, min(len(files), 10)):
                f_path = os.path.join(p, sub, f)
                img = extract_frame(f_path)
                spec, txt_a = extract_audio(f_path)
                seq = text_to_sequence(txt_a)
                res = model.predict([np.expand_dims(img,0), np.expand_dims(spec,0), np.expand_dims(seq,0)], verbose=0)[0][0]
                print(f"[{'Kasar' if res >=0.5 else 'Non-Kasar'}] {res*100:.2f}% | {f[:30]}...")

if __name__ == "__main__":
    main()
```
