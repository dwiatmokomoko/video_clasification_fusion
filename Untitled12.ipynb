```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings
import tensorflow as tf
import librosa, whisper, easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns
warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data.npz"
MODEL_FILE = "best_model.h5"

IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 20

# ===============================
# LOAD MODELS
# ===============================
print("Loading Whisper and OCR...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id','en'], gpu=False)

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total//2)
    ret, frame = cap.read()
    cap.release()
    if not ret:
        return np.zeros((*IMG_SIZE,3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame/255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = "temp.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None:
                return np.zeros((*SPEC_SIZE,1)), ""
            video.audio.write_audiofile(tmp, logger=None)

        y, sr = librosa.load(tmp, sr=SR)

        if len(y) < SR*DURATION:
            y = np.pad(y, (0, SR*DURATION-len(y)))

        y = y[:SR*DURATION] + 0.005*np.random.randn(SR*DURATION)

        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel))/(np.std(mel)+1e-9)

        text = ""
        try:
            text = stt_model.transcribe(tmp)['text']
        except:
            pass

        os.remove(tmp)
        return mel.astype(np.float32), text.lower()

    except Exception as e:
        return np.zeros((*SPEC_SIZE,1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame*255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except:
        return ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c)%5000 for c in text][:MAX_TEXT_LEN]
    seq += [0]*(MAX_TEXT_LEN-len(seq))
    return np.array(seq)

# ===============================
# LOAD DATA (RECURSIVE)
# ===============================
def load_or_extract_data():
    if os.path.exists(PROCESSED_DATA_FILE):
        print("Loading cached dataset...")
        data = np.load(PROCESSED_DATA_FILE)
        return data['X_img'], data['X_spec'], data['X_txt'], data['y']

    X_img, X_spec, X_txt, y = [], [], [], []

    for main_folder in os.listdir(BASE_DATASET):
        main_path = os.path.join(BASE_DATASET, main_folder)
        if not os.path.isdir(main_path):
            continue

        if main_folder.lower() == "nonkasar":
            label = 0
        elif main_folder.lower() == "kasar":
            label = 1
        else:
            continue

        print(f"\nProcessing {main_folder} (label={label})")

        for root, _, files in os.walk(main_path):
            for f in files:
                if not f.lower().endswith((".mp4",".avi",".mov")):
                    continue

                full = os.path.join(root, f)

                try:
                    frame = extract_frame(full)
                    spec, text_audio = extract_audio(full)
                    text_ocr = extract_text_ocr(frame)
                    txt_seq = text_to_sequence(text_audio + " " + text_ocr)

                    X_img.append(frame)
                    X_spec.append(spec)
                    X_txt.append(txt_seq)
                    y.append(label)

                except Exception as e:
                    print("ERROR:", full, "|", e)

    X_img = np.array(X_img)
    X_spec = np.array(X_spec)
    X_txt = np.array(X_txt)
    y = np.array(y)

    print("\nTotal data:", len(y))

    if len(y) == 0:
        raise ValueError("Dataset kosong! Cek struktur folder dan file.")

    np.savez(PROCESSED_DATA_FILE, X_img=X_img, X_spec=X_spec, X_txt=X_txt, y=y)
    return X_img, X_spec, X_txt, y

# ===============================
# CLASS WEIGHT
# ===============================
def get_class_weights(y):
    classes = np.unique(y)
    weights = compute_class_weight('balanced', classes=classes, y=y)
    class_weight = dict(zip(classes, weights))

    print("Distribusi:", np.bincount(y))
    print("Class weight:", class_weight)

    return class_weight

# ===============================
# MODEL
# ===============================
from tensorflow.keras import backend as K

class AdaptiveFusion(layers.Layer):
    def __init__(self):
        super().__init__()
        self.dense = layers.Dense(3, activation='softmax')

    def call(self, inputs):
        v,a,t = inputs
        w = self.dense(layers.concatenate([v,a,t]))
        v *= K.expand_dims(w[:,0],-1)
        a *= K.expand_dims(w[:,1],-1)
        t *= K.expand_dims(w[:,2],-1)
        return layers.concatenate([v,a,t])

def build_model():
    img_in = layers.Input(shape=(*IMG_SIZE,3))
    spec_in = layers.Input(shape=(*SPEC_SIZE,1))
    txt_in = layers.Input(shape=(MAX_TEXT_LEN,))

    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    base.trainable = False

    v = layers.GlobalAveragePooling2D()(base.output)

    a = layers.Conv2D(32,3,activation='relu',padding='same')(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64,3,activation='relu',padding='same')(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((1024,64))(a)
    a = layers.Bidirectional(layers.GRU(64))(a)

    t = layers.Embedding(5000,128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64))(t)

    fused = AdaptiveFusion()([v,a,t])

    x = layers.Dense(256,activation='relu')(fused)
    x = layers.Dropout(0.5)(x)
    out = layers.Dense(1,activation='sigmoid')(x)

    model = models.Model([img_in,spec_in,txt_in], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ===============================
# MAIN
# ===============================
def main():
    # ⚠️ HAPUS CACHE LAMA OTOMATIS
    if os.path.exists(PROCESSED_DATA_FILE):
        os.remove(PROCESSED_DATA_FILE)

    X_img, X_spec, X_txt, y = load_or_extract_data()

    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_txt_tr, X_txt_val, y_tr, y_val = train_test_split(
        X_img, X_spec, X_txt, y,
        test_size=0.2,
        stratify=y,
        random_state=42
    )

    class_weight = get_class_weights(y_tr)

    model = build_model()

    model.fit(
        [X_img_tr,X_spec_tr,X_txt_tr],
        y_tr,
        validation_data=([X_img_val,X_spec_val,X_txt_val], y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=class_weight
    )

    preds = model.predict([X_img_val,X_spec_val,X_txt_val])
    preds_class = (preds > 0.5).astype(int)

    print(classification_report(y_val, preds_class))

if __name__ == "__main__":
    main()
```

```python
import random
import os
import numpy as np
import tensorflow as tf

# ============================
# PREDICT 1 FILE
# ============================
def predict_file(file_path, model):
    frame = extract_frame(file_path)
    spec, text_audio = extract_audio(file_path)
    text_ocr = extract_text_ocr(frame)
    txt_seq = text_to_sequence(text_audio + " " + text_ocr)

    pred = model.predict([
        np.expand_dims(frame, axis=0),
        np.expand_dims(spec, axis=0),
        np.expand_dims(txt_seq, axis=0)
    ], verbose=0)

    prob = float(pred[0][0])
    pred_class = 1 if prob > 0.5 else 0
    category = "Kasar" if pred_class == 1 else "Non Kasar"

    return pred_class, category, prob


# ============================
# AMBIL SEMUA VIDEO (RECURSIVE)
# ============================
def get_all_video_files(folder_path):
    video_files = []
    for root, _, files in os.walk(folder_path):
        for f in files:
            if f.lower().endswith((".mp4", ".avi", ".mov")):
                video_files.append(os.path.join(root, f))
    return video_files


# ============================
# PREDICT + HITUNG ACCURACY
# ============================
def evaluate_folder(folder_path, model, true_label, num_files=10):
    files = get_all_video_files(folder_path)

    if len(files) == 0:
        print(f"Tidak ada video di: {folder_path}")
        return

    random_files = random.sample(files, min(num_files, len(files)))

    correct = 0
    total = 0

    print(f"\n===== HASIL PREDIKSI: {folder_path} =====")

    for file_path in random_files:
        pred_class, category, confidence = predict_file(file_path, model)

        # Hitung benar/salah
        if pred_class == true_label:
            correct += 1

        total += 1

        print(f"\nFile: {file_path}")
        print(f"Prediksi   : {category}")
        print(f"Confidence : {confidence:.4f}")

    accuracy = correct / total if total > 0 else 0

    print("\n-----------------------------")
    print(f"Total Data : {total}")
    print(f"Benar      : {correct}")
    print(f"Accuracy   : {accuracy:.4f}")
    print("-----------------------------")

    return accuracy


# ============================
# LOAD MODEL & TEST
# ============================
if os.path.exists(MODEL_FILE):
    model = tf.keras.models.load_model(
        MODEL_FILE,
        custom_objects={'AdaptiveFusion': AdaptiveFusion}
    )

    dataset_path = r"D:\web\cnn_dnn_rnn_gan\Dataset"

    # Label sesuai training:
    # Nonkasar = 0
    # Kasar = 1

    acc_nonkasar = evaluate_folder(
        os.path.join(dataset_path, "Nonkasar"),
        model,
        true_label=0
    )

    acc_kasar = evaluate_folder(
        os.path.join(dataset_path, "Kasar"),
        model,
        true_label=1
    )

    # 🔥 TOTAL ACCURACY
    if acc_nonkasar is not None and acc_kasar is not None:
        total_acc = (acc_nonkasar + acc_kasar) / 2
        print(f"\n🔥 Rata-rata Accuracy: {total_acc:.4f}")

else:
    print("Model belum ditemukan!")
```
