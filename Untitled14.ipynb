```python
# -*- coding: utf-8 -*-
import os
import re
import random
import cv2
import numpy as np
import warnings
import tempfile
import tensorflow as tf
import librosa
import whisper
import easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip

from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns

warnings.filterwarnings("ignore")

# ===============================
# REPRODUCIBILITY
# ===============================
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data_final.npz"
MODEL_FILE = "best_model.keras"
HISTORY_FILE = "history_final.npz"
IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
MAX_TEXT_LEN = 100
MAX_VOCAB = 10000
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS_STAGE1 = 20
EPOCHS_STAGE2 = 10
CACHE_VERSION = "kasar_nonkasar_final_v1"

VIDEO_EXTENSIONS = (".mp4", ".avi", ".mov", ".mkv", ".webm")

# Folder label utama
CLASS_MAP = {
    "kasar": 1,
    "nonkasar": 0
}

# ===============================
# LOAD MODELS
# ===============================
print("Loading Whisper and OCR models...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=False)

# ===============================
# UTILITIES
# ===============================
def normalize_label(folder_name):
    """
    Mengubah:
    - Kasar -> kasar
    - Nonkasar -> nonkasar
    - Non Kasar -> nonkasar
    - non_kasar -> nonkasar
    """
    name = folder_name.strip().lower()
    name = re.sub(r"[\s_\-]+", "", name)
    return name

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    text = text.strip()
    return text

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total <= 0:
        cap.release()
        return np.zeros((*IMG_SIZE, 3), dtype=np.float32)

    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()

    if not ret or frame is None:
        return np.zeros((*IMG_SIZE, 3), dtype=np.float32)

    frame = cv2.resize(frame, IMG_SIZE)
    frame = frame.astype(np.float32)
    return frame

def extract_audio(video_path):
    tmp_wav = None
    try:
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
            tmp_wav = tmp_file.name

        with VideoFileClip(video_path) as video:
            if video.audio is None:
                if tmp_wav and os.path.exists(tmp_wav):
                    os.remove(tmp_wav)
                return np.zeros((*SPEC_SIZE, 1), dtype=np.float32), ""

            video.audio.write_audiofile(tmp_wav, logger=None)

        y, sr = librosa.load(tmp_wav, sr=SR)

        target_len = SR * DURATION
        if len(y) < target_len:
            y = np.pad(y, (0, target_len - len(y)))
        else:
            y = y[:target_len]

        y = y + 0.005 * np.random.randn(len(y))

        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel_db = librosa.power_to_db(mel, ref=np.max)
        mel_db = cv2.resize(mel_db, SPEC_SIZE)
        mel_db = np.expand_dims(mel_db, -1)
        mel_db = (mel_db - np.mean(mel_db)) / (np.std(mel_db) + 1e-9)

        text = ""
        try:
            text = stt_model.transcribe(tmp_wav)['text']
        except:
            text = ""

        if tmp_wav and os.path.exists(tmp_wav):
            os.remove(tmp_wav)

        return mel_db.astype(np.float32), clean_text(text)

    except:
        if tmp_wav and os.path.exists(tmp_wav):
            try:
                os.remove(tmp_wav)
            except:
                pass
        return np.zeros((*SPEC_SIZE, 1), dtype=np.float32), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = np.clip(frame, 0, 255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return clean_text(text)
    except:
        return ""

# ===============================
# LOAD / CACHE DATA
# ===============================
def load_or_extract_data():
    if os.path.exists(PROCESSED_DATA_FILE):
        try:
            data = np.load(PROCESSED_DATA_FILE, allow_pickle=True)
            cache_version = str(data["cache_version"].item())
            if cache_version == CACHE_VERSION:
                X_img = data["X_img"].astype(np.float32)
                X_spec = data["X_spec"].astype(np.float32)
                X_text = data["X_text"].astype(str)
                y = data["y"].astype(np.int32)
                print("Loaded cached dataset.")
                return X_img, X_spec, X_text, y
            else:
                print("Cache versi lama ditemukan, ekstraksi ulang dataset...")
        except:
            print("Cache rusak atau tidak cocok, ekstraksi ulang dataset...")

    X_img, X_spec, X_text, y = [], [], [], []
    count_per_class = {0: 0, 1: 0}

    if not os.path.exists(BASE_DATASET):
        raise FileNotFoundError(f"Folder dataset tidak ditemukan: {BASE_DATASET}")

    for class_folder in os.listdir(BASE_DATASET):
        class_path = os.path.join(BASE_DATASET, class_folder)
        if not os.path.isdir(class_path):
            continue

        norm_folder = normalize_label(class_folder)
        if norm_folder not in CLASS_MAP:
            print(f"Ignored folder: {class_folder}")
            continue

        label = CLASS_MAP[norm_folder]
        print(f"Reading folder: {class_folder} -> label {label}")

        for root, _, files in os.walk(class_path):
            for f in files:
                if not f.lower().endswith(VIDEO_EXTENSIONS):
                    continue

                full_path = os.path.join(root, f)

                try:
                    frame = extract_frame(full_path)
                    spec, text_audio = extract_audio(full_path)
                    text_ocr = extract_text_ocr(frame)
                    combined_text = clean_text(f"{text_audio} {text_ocr}")

                    X_img.append(frame)
                    X_spec.append(spec)
                    X_text.append(combined_text)
                    y.append(label)
                    count_per_class[label] += 1

                    print(f"Processed: {full_path}")

                except Exception as e:
                    print(f"Skip: {full_path} | Error: {e}")

    print("Jumlah data kelas 0 (Nonkasar):", count_per_class[0])
    print("Jumlah data kelas 1 (Kasar):", count_per_class[1])

    if len(X_img) == 0:
        raise ValueError("Tidak ada video yang berhasil diproses. Periksa struktur folder dataset.")

    X_img = np.array(X_img, dtype=np.float32)
    X_spec = np.array(X_spec, dtype=np.float32)
    X_text = np.array(X_text, dtype=str)
    y = np.array(y, dtype=np.int32)

    np.savez(
        PROCESSED_DATA_FILE,
        X_img=X_img,
        X_spec=X_spec,
        X_text=X_text,
        y=y,
        cache_version=np.array(CACHE_VERSION)
    )

    print("Dataset berhasil diekstrak dan disimpan ke cache.")
    return X_img, X_spec, X_text, y

# ===============================
# ADAPTIVE FUSION LAYER
# ===============================
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.gate = layers.Dense(3, activation='softmax')

    def call(self, inputs):
        v, a, t = inputs
        combined = tf.concat([v, a, t], axis=-1)
        weights = self.gate(combined)

        wv = tf.expand_dims(weights[:, 0], -1)
        wa = tf.expand_dims(weights[:, 1], -1)
        wt = tf.expand_dims(weights[:, 2], -1)

        v = v * wv
        a = a * wa
        t = t * wt

        return tf.concat([v, a, t], axis=-1)

# ===============================
# BUILD MODEL
# ===============================
def build_model(text_vectorizer):
    img_in = layers.Input(shape=(*IMG_SIZE, 3), name="image_input")
    spec_in = layers.Input(shape=(*SPEC_SIZE, 1), name="spec_input")
    txt_in = layers.Input(shape=(), dtype=tf.string, name="text_input")

    # Augmentasi gambar
    img_aug = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.05),
        layers.RandomZoom(0.05),
        layers.RandomContrast(0.1),
    ], name="image_augmentation")

    x_img = img_aug(img_in)

    # Visual branch
    base = EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_tensor=x_img
    )
    base.trainable = False

    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(v)
    v = layers.BatchNormalization()(v)
    v = layers.Dropout(0.4)(v)
    v = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(v)

    # Audio branch
    a = layers.Conv2D(32, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(spec_in)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Dropout(0.2)(a)

    a = layers.Conv2D(64, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Dropout(0.2)(a)

    a = layers.Conv2D(128, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)

    a = layers.Reshape((1024, 128))(a)
    a = layers.Bidirectional(layers.GRU(64, dropout=0.2, recurrent_dropout=0.1))(a)
    a = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.Dropout(0.3)(a)

    # Text branch
    t = text_vectorizer(txt_in)
    t = layers.Embedding(MAX_VOCAB, 128, mask_zero=True)(t)
    t = layers.Bidirectional(layers.GRU(64, dropout=0.2, recurrent_dropout=0.1))(t)
    t = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(t)
    t = layers.Dropout(0.3)(t)

    # Fusion
    fused = AdaptiveFusion()([v, a, t])
    fused = layers.LayerNormalization()(fused)

    # Classifier
    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation='sigmoid', dtype='float32')(x)

    model = models.Model([img_in, spec_in, txt_in], out)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model, base

def unfreeze_backbone(base_model, fine_tune_at=300):
    base_model.trainable = True

    for layer in base_model.layers[:fine_tune_at]:
        layer.trainable = False

    for layer in base_model.layers[fine_tune_at:]:
        if not isinstance(layer, layers.BatchNormalization):
            layer.trainable = True

# ===============================
# VISUALIZATION
# ===============================
def visualize_model_architecture(model):
    from tensorflow.keras.utils import plot_model
    plot_model(model, to_file="model_architecture.png", show_shapes=True, show_layer_names=True)
    print("Model architecture saved as model_architecture.png")

def plot_training_history(history_dict):
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history_dict.get('accuracy', []), label='Train Accuracy')
    plt.plot(history_dict.get('val_accuracy', []), label='Val Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history_dict.get('loss', []), label='Train Loss')
    plt.plot(history_dict.get('val_loss', []), label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=['Non Kasar', 'Kasar'],
        yticklabels=['Non Kasar', 'Kasar']
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()

def merge_histories(h1, h2):
    if h1 is None:
        return h2.copy()
    if h2 is None:
        return h1.copy()

    merged = {}
    keys = set(h1.keys()).union(h2.keys())
    for k in keys:
        v1 = h1.get(k, [])
        v2 = h2.get(k, [])
        merged[k] = list(v1) + list(v2)
    return merged

# ===============================
# PREDICTION HELPERS
# ===============================
def predict_file(file_path, model):
    frame = extract_frame(file_path)
    spec, text_audio = extract_audio(file_path)
    text_ocr = extract_text_ocr(frame)
    combined_text = clean_text(f"{text_audio} {text_ocr}")

    img_input = np.expand_dims(frame, axis=0).astype(np.float32)
    spec_input = np.expand_dims(spec, axis=0).astype(np.float32)
    text_input = np.array([combined_text], dtype=str)

    pred = model.predict([img_input, spec_input, text_input], verbose=0)
    prob_kasar = float(pred[0][0])

    pred_class = 1 if prob_kasar >= 0.5 else 0
    category = "Kasar" if pred_class == 1 else "Non Kasar"

    return category, prob_kasar

def get_video_files_recursive(folder_path):
    video_files = []
    for root, _, files in os.walk(folder_path):
        for f in files:
            if f.lower().endswith(VIDEO_EXTENSIONS):
                video_files.append(os.path.join(root, f))
    return video_files

def predict_folder(folder_path, model, num_files=10):
    results = {}

    if not os.path.exists(folder_path):
        print(f"Folder tidak ditemukan: {folder_path}")
        return results

    files = get_video_files_recursive(folder_path)
    if len(files) == 0:
        print(f"Tidak ada file video di: {folder_path}")
        return results

    chosen_files = random.sample(files, num_files) if len(files) > num_files else files

    for file_path in chosen_files:
        file_name = os.path.basename(file_path)
        try:
            category, confidence = predict_file(file_path, model)
            results[file_name] = {
                "file_path": file_path,
                "category": category,
                "confidence": confidence
            }
        except Exception as e:
            results[file_name] = {
                "file_path": file_path,
                "category": "Error",
                "confidence": None,
                "error": str(e)
            }

    return results

# ===============================
# MAIN TRAINING
# ===============================
def main_train():
    X_img, X_spec, X_text, y = load_or_extract_data()

    unique_classes = np.unique(y)
    print("Unique classes:", unique_classes)

    if len(unique_classes) < 2:
        raise ValueError(
            "Dataset masih hanya berisi satu kelas. "
            "Pastikan folder Kasar dan Nonkasar keduanya berisi video."
        )

    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_text_tr, X_text_val, y_tr, y_val = train_test_split(
        X_img,
        X_spec,
        X_text,
        y,
        test_size=0.2,
        stratify=y,
        random_state=SEED
    )

    # Class Weight
    weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_tr),
        y=y_tr
    )
    class_weight = dict(zip(np.unique(y_tr), weights))
    print("Class weight:", class_weight)

    # Text vectorizer
    text_vectorizer = layers.TextVectorization(
        max_tokens=MAX_VOCAB,
        output_mode='int',
        output_sequence_length=MAX_TEXT_LEN,
        standardize='lower_and_strip_punctuation',
        name='text_vectorizer'
    )
    text_vectorizer.adapt(X_text_tr)

    if os.path.exists(MODEL_FILE):
        model = tf.keras.models.load_model(
            MODEL_FILE,
            custom_objects={'AdaptiveFusion': AdaptiveFusion}
        )
        history_all = None
        print(f"Loaded model from {MODEL_FILE}")
    else:
        model, base_model = build_model(text_vectorizer)

        callbacks_stage1 = [
            tf.keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=5,
                restore_best_weights=True
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=2,
                min_lr=1e-6
            ),
            tf.keras.callbacks.ModelCheckpoint(
                MODEL_FILE,
                monitor='val_loss',
                save_best_only=True
            ),
            tf.keras.callbacks.TerminateOnNaN()
        ]

        history1 = model.fit(
            [X_img_tr, X_spec_tr, X_text_tr],
            y_tr,
            validation_data=([X_img_val, X_spec_val, X_text_val], y_val),
            epochs=EPOCHS_STAGE1,
            batch_size=BATCH_SIZE,
            callbacks=callbacks_stage1,
            class_weight=class_weight,
            shuffle=True
        )

        # Fine-tuning backbone
        unfreeze_backbone(base_model, fine_tune_at=max(0, len(base_model.layers) - 30))

        model.compile(
            optimizer=tf.keras.optimizers.Adam(1e-5),
            loss='binary_crossentropy',
            metrics=[
                'accuracy',
                tf.keras.metrics.AUC(name='auc'),
                tf.keras.metrics.Precision(name='precision'),
                tf.keras.metrics.Recall(name='recall')
            ]
        )

        callbacks_stage2 = [
            tf.keras.callbacks.EarlyStopping(
                monitor='val_loss',
                patience=4,
                restore_best_weights=True
            ),
            tf.keras.callbacks.ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=2,
                min_lr=1e-7
            ),
            tf.keras.callbacks.ModelCheckpoint(
                MODEL_FILE,
                monitor='val_loss',
                save_best_only=True
            ),
            tf.keras.callbacks.TerminateOnNaN()
        ]

        history2 = model.fit(
            [X_img_tr, X_spec_tr, X_text_tr],
            y_tr,
            validation_data=([X_img_val, X_spec_val, X_text_val], y_val),
            epochs=EPOCHS_STAGE2,
            batch_size=BATCH_SIZE,
            callbacks=callbacks_stage2,
            class_weight=class_weight,
            shuffle=True
        )

        history_all = merge_histories(history1.history, history2.history)
        np.savez(
            HISTORY_FILE,
            **{k: np.array(v) for k, v in history_all.items()}
        )
        print(f"Training history saved to {HISTORY_FILE}")

    visualize_model_architecture(model)

    preds = model.predict([X_img_val, X_spec_val, X_text_val], batch_size=BATCH_SIZE, verbose=0)
    preds_class = (preds > 0.5).astype(int).flatten()

    print(classification_report(
        y_val,
        preds_class,
        target_names=['Non Kasar', 'Kasar'],
        digits=4
    ))

    plot_confusion_matrix(y_val, preds_class)

    if 'history_all' in locals() and history_all is not None:
        plot_training_history(history_all)

# ===============================
# MAIN PREDICTION
# ===============================
def main_predict():
    if not os.path.exists(MODEL_FILE):
        print("Model belum ditemukan, pastikan model sudah dilatih dan disimpan.")
        return

    model = tf.keras.models.load_model(
        MODEL_FILE,
        custom_objects={'AdaptiveFusion': AdaptiveFusion}
    )

    folders = ["Kasar", "Nonkasar"]

    for folder in folders:
        folder_path = os.path.join(BASE_DATASET, folder)
        results = predict_folder(folder_path, model, num_files=10)

        print(f"\nHasil Prediksi untuk Folder '{folder}':")
        for file_name, result in results.items():
            print(f"File: {result['file_path']}")
            print(f"  Kategori: {result['category']}")
            if result.get("confidence") is not None:
                print(f"  Confidence Kasar: {result['confidence']:.4f}")
            if "error" in result:
                print(f"  Error: {result['error']}")

# ===============================
# RUN
# ===============================
if __name__ == "__main__":
    # Jalankan training
    main_train()

    # Setelah training selesai, jika ingin langsung uji prediksi folder,
    # aktifkan baris di bawah.
    # main_predict()
```
