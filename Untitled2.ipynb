```python
import os
import cv2
import librosa
import whisper
import easyocr
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import tensorflow as tf
from tensorflow.keras import layers, models
from moviepy import VideoFileClip
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ==========================================
# 1. KONFIGURASI & KAMUS KATA KASAR
# ==========================================
BASE_PATH = 'D:/web/cnn_dnn_rnn_gan/dataset'
SAVE_DATA_PATH = 'dataset_final_robust.pkl'
IMG_SIZE = (128, 128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 2.0

# Daftar Slang & Kata Kasar Lokal (Dictionary)
KATA_KASAR_INDO = [
    'anjing', 'babi', 'monyet', 'bajingan', 'asu', 'bangsat', 'kampret',
    'goblok', 'tolol', 'idiot', 'bego', 'sinting', 'jancok', 'jancuk', 
    'pukimak', 'lontre', 'tai', 'ngentot', 'memek', 'kontol', 'pantek'
]

print("Memuat AI Models (Whisper & OCR)...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'])

# ==========================================
# 2. FUNGSI PEMROSESAN AUDIO (ROBUST)
# ==========================================
def safe_trim_pad(y, sr, duration):
    try:
        y_trimmed, _ = librosa.effects.trim(y, top_db=25)
        y = y_trimmed if len(y_trimmed) > 0 else y
    except: pass
    target_len = int(sr * duration)
    return np.pad(y, (0, max(0, target_len - len(y))), mode='constant')[:target_len]

def extract_mel_spec(y, sr):
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    # Normalisasi ke 0-1 untuk stabilitas CNN
    mel_db = (mel_db - mel_db.min()) / (mel_db.max() - mel_db.min() + 1e-8)
    mel_resize = cv2.resize(mel_db, IMG_SIZE)
    return np.expand_dims(mel_resize, axis=-1)

# ==========================================
# 3. EKSTRAKSI FITUR MULTIMODAL
# ==========================================
def extract_features():
    if os.path.exists(SAVE_DATA_PATH):
        print("Memuat cache fitur...")
        with open(SAVE_DATA_PATH, 'rb') as f: return pickle.load(f)

    data_list = []
    for folder in os.listdir(BASE_PATH):
        f_path = os.path.join(BASE_PATH, folder)
        if not os.path.isdir(f_path): continue
        label = 0 if folder.lower() == 'aman' else 1
        
        for v_file in [f for f in os.listdir(f_path) if f.endswith(('.mp4', '.mkv'))]:
            path = os.path.join(f_path, v_file)
            try:
                # Visual
                cap = cv2.VideoCapture(path)
                cap.set(cv2.CAP_PROP_POS_FRAMES, int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) // 2)
                ret, frame = cap.read()
                img = cv2.resize(cv2.convertScaleAbs(frame), IMG_SIZE).astype(np.float32)/255.0 if ret else np.zeros((*IMG_SIZE, 3))
                cap.release()

                # Audio & Text
                tmp_a = "temp_proc.wav"
                with VideoFileClip(path) as video:
                    video.audio.write_audiofile(tmp_a, logger=None)
                
                y_raw, sr = librosa.load(tmp_a, sr=SR)
                y_proc = safe_trim_pad(y_raw, SR, DURATION)
                spec = extract_mel_spec(y_proc, SR)
                
                # Transkrip + OCR
                text_stt = stt_model.transcribe(tmp_a)['text']
                text_ocr = " ".join(ocr_reader.readtext(frame if ret else np.zeros(IMG_SIZE), detail=0))
                full_text = (text_stt + " " + text_ocr).lower()
                full_text = re.sub(r'(.)\1+', r'\1', full_text) # Normalisasi typo (goblokkk -> goblok)

                data_list.append({'img': img, 'spec': spec, 'text': full_text, 'label': label})
                if os.path.exists(tmp_a): os.remove(tmp_a)
                print(f"✅ Berhasil: {v_file}")
            except Exception as e: print(f"❌ Error {v_file}: {e}")

    with open(SAVE_DATA_PATH, 'wb') as f: pickle.dump(data_list, f)
    return data_list

# ==========================================
# 4. ARSITEKTUR MODEL (CNN-RNN FUSION)
# ==========================================
def build_model():
    in_img = layers.Input(shape=(128, 128, 3))
    in_spec = layers.Input(shape=(128, 128, 1))
    in_txt = layers.Input(shape=(MAX_TEXT_LEN,))

    # Visual Branch
    v = layers.Conv2D(32, (3,3), activation='relu')(in_img)
    v = layers.MaxPooling2D()(v)
    v = layers.Flatten()(v)

    # Audio Branch
    a = layers.Conv2D(32, (3,3), activation='relu')(in_spec)
    a = layers.MaxPooling2D()(a)
    a = layers.Flatten()(a)

    # Text Branch (RNN)
    t = layers.Embedding(5000, 64)(in_txt)
    t = layers.LSTM(64)(t)

    # Late Fusion
    m = layers.concatenate([v, a, t])
    d = layers.Dense(128, activation='relu')(m)
    d = layers.Dropout(0.5)(d)
    out = layers.Dense(1, activation='sigmoid')(d)

    model = models.Model(inputs=[in_img, in_spec, in_txt], outputs=out)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ==========================================
# 5. VISUALISASI HASIL & DATA
# ==========================================
def plot_results(history, y_ts, y_pred):
    plt.figure(figsize=(15, 5))
    
    # 1. Training Curve
    plt.subplot(1, 3, 1)
    plt.plot(history.history['accuracy'], label='Train Acc')
    plt.plot(history.history['val_accuracy'], label='Val Acc')
    plt.title('Accuracy History')
    plt.legend()

    # 2. Confusion Matrix
    plt.subplot(1, 3, 2)
    cm = confusion_matrix(y_ts, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')

    # 3. Contoh Melspectrogram
    plt.subplot(1, 3, 3)
    plt.imshow(X_ts_spec[0].squeeze(), cmap='magma')
    plt.title('Sample Mel-Spectrogram')
    plt.show()

# ==========================================
# 6. INFERENCE (TES VIDEO BARU)
# ==========================================
def run_inference(v_path, model, tokenizer):
    print(f"\nPrediksi Video: {v_path}")
    # (Logika ekstraksi sama dengan extract_features untuk 1 file)
    # ... [Ekstraksi frame, spec, text] ...
    # Prediksi menggunakan model.predict()
    print("Hasil: AMAN / KASAR")

# ==========================================
# RUN PROCESS
# ==========================================
if __name__ == "__main__":
    data = extract_features()
    df = pd.DataFrame(data)

    # Balancing
    df_aman = df[df['label'] == 0]
    df_kasar = df[df['label'] == 1].sample(len(df_aman))
    df = pd.concat([df_aman, df_kasar]).sample(frac=1)

    # Tokenizer + Dictionary
    tokenizer = tf.keras.preprocessing.text.Tokenizer(num_words=5000, oov_token="<OOV>")
    tokenizer.fit_on_texts(df['text'].tolist() + KATA_KASAR_INDO)
    X_txt = tf.keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences(df['text']), maxlen=MAX_TEXT_LEN)
    
    X_img, X_spec, y = np.array(df['img'].tolist()), np.array(df['spec'].tolist()), df['label'].values
    
    X_tr_img, X_ts_img, X_tr_spec, X_ts_spec, X_tr_txt, X_ts_txt, y_tr, y_ts = train_test_split(X_img, X_spec, X_txt, y, test_size=0.2)

    model = build_model()
    history = model.fit([X_tr_img, X_tr_spec, X_tr_txt], y_tr, 
                        validation_data=([X_ts_img, X_ts_spec, X_ts_txt], y_ts),
                        epochs=15, batch_size=8)

    y_pred = (model.predict([X_ts_img, X_ts_spec, X_ts_txt]) > 0.5).astype(int)
    plot_results(history, y_ts, y_pred)
```
