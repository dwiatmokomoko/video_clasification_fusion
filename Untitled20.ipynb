```python
# -*- coding: utf-8 -*-
import os
import sys
import re
import random
import cv2
import numpy as np
import warnings
import tempfile
import logging

import tensorflow as tf
import librosa
import whisper
import easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tqdm.auto import tqdm

from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import seaborn as sns

warnings.filterwarnings("ignore")
logging.getLogger("tensorflow").setLevel(logging.ERROR)

# ===============================
# UTF-8 SAFETY FIX UNTUK WINDOWS
# ===============================
os.environ["PYTHONUTF8"] = "1"
os.environ["PYTHONIOENCODING"] = "utf-8"

# Memaksa stdout dan stderr menggunakan UTF-8 saat runtime
if hasattr(sys.stdout, 'reconfigure'):
    sys.stdout.reconfigure(encoding='utf-8')
if hasattr(sys.stderr, 'reconfigure'):
    sys.stderr.reconfigure(encoding='utf-8')

# ===============================
# REPRODUCIBILITY
# ===============================
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
random.seed(SEED)

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"

# Cache lama: dipakai hanya jika formatnya kompatibel
PROCESSED_DATA_FILE = "processed_data_final.npz"

# Cache ringan baru
MANIFEST_CACHE_FILE = "manifest_flowchart.npz"

MODEL_FILE = "best_model_flowchart.keras"
HISTORY_FILE = "history_flowchart.npz"

IMG_SIZE = (224, 224)
SPEC_SIZE = (128, 128)
NUM_FRAMES = 8

MAX_TEXT_LEN = 120
MAX_VOCAB = 10000

SR = 22050
DURATION = 3

BATCH_SIZE = 4
EPOCHS_STAGE1 = 20
EPOCHS_STAGE2 = 10

CACHE_VERSION = "flowchart_manifest_v3"

VIDEO_EXTENSIONS = (".mp4", ".avi", ".mov", ".mkv", ".webm")

CLASS_MAP = {
    "kasar": 1,
    "nonkasar": 0
}

# ===============================
# LOAD MODELS
# ===============================
print("Loading Whisper and OCR models...")
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id', 'en'], gpu=False)

# ===============================
# UTILITIES
# ===============================
def normalize_label(folder_name):
    name = folder_name.strip().lower()
    name = re.sub(r"[\s_\-]+", "", name)
    return name

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    text = text.strip()
    return text

def safe_remove(path):
    try:
        if path and os.path.exists(path):
            os.remove(path)
    except:
        pass

def format_float(v):
    try:
        return f"{float(v):.4f}"
    except:
        return str(v)

# ===============================
# CACHE CHECKS
# ===============================
def load_manifest_if_compatible(npz_path):
    if not os.path.exists(npz_path):
        return None

    try:
        data = np.load(npz_path, allow_pickle=True)
        files = set(data.files)

        required = {"video_paths", "labels", "texts", "specs", "cache_version"}
        if not required.issubset(files):
            return None

        cache_version = str(data["cache_version"].item())
        if cache_version != CACHE_VERSION:
            return None

        video_paths = data["video_paths"].astype(str)
        labels = data["labels"].astype(np.int32)
        texts = data["texts"].astype(str)
        specs = data["specs"]

        if not (len(video_paths) == len(labels) == len(texts) == len(specs)):
            return None

        return {
            "video_paths": video_paths,
            "labels": labels,
            "texts": texts,
            "specs": specs
        }

    except:
        return None

def try_use_processed_data_file():
    """
    processed_data_final.npz dipakai hanya kalau ternyata isinya manifest ringan.
    Kalau berisi X_video / tensor besar, file diabaikan supaya tidak MemoryError.
    """
    if not os.path.exists(PROCESSED_DATA_FILE):
        return None

    try:
        data = np.load(PROCESSED_DATA_FILE, allow_pickle=True)
        files = set(data.files)

        # Kasus kompatibel: ternyata sudah manifest ringan
        if {"video_paths", "labels", "texts", "specs", "cache_version"}.issubset(files):
            cache_version = str(data["cache_version"].item())
            if cache_version == CACHE_VERSION:
                return {
                    "video_paths": data["video_paths"].astype(str),
                    "labels": data["labels"].astype(np.int32),
                    "texts": data["texts"].astype(str),
                    "specs": data["specs"]
                }

        # Kasus lama: tensor penuh, sengaja diabaikan
        if {"X_video", "X_spec", "X_text", "y"}.issubset(files):
            return None

        return None

    except:
        return None

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frames(video_path, num_frames=NUM_FRAMES):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total <= 0:
        cap.release()
        return np.zeros((num_frames, IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.float32)

    indices = np.linspace(0, total - 1, num_frames).astype(int)
    frames = []

    for idx in indices:
        cap.set(cv2.CAP_PROP_POS_FRAMES, int(idx))
        ret, frame = cap.read()

        if not ret or frame is None:
            if len(frames) > 0:
                frames.append(frames[-1].copy())
            else:
                frames.append(np.zeros((IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.float32))
            continue

        frame = cv2.resize(frame, IMG_SIZE)
        frames.append(frame.astype(np.float32))

    cap.release()

    while len(frames) < num_frames:
        frames.append(frames[-1].copy() if len(frames) > 0 else np.zeros((IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.float32))

    return np.array(frames[:num_frames], dtype=np.float32)

def extract_middle_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    if total <= 0:
        cap.release()
        return np.zeros((IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.float32)

    cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()

    if not ret or frame is None:
        return np.zeros((IMG_SIZE[1], IMG_SIZE[0], 3), dtype=np.float32)

    frame = cv2.resize(frame, IMG_SIZE)
    return frame.astype(np.float32)

def extract_audio(video_path):
    tmp_wav = None
    try:
        with tempfile.NamedTemporaryFile(suffix=".wav", delete=False) as tmp_file:
            tmp_wav = tmp_file.name

        with VideoFileClip(video_path) as video:
            if video.audio is None:
                safe_remove(tmp_wav)
                return np.zeros((SPEC_SIZE[1], SPEC_SIZE[0], 1), dtype=np.float32), ""

            video.audio.write_audiofile(tmp_wav, logger=None)

        y, sr = librosa.load(tmp_wav, sr=SR)

        target_len = SR * DURATION
        if len(y) < target_len:
            y = np.pad(y, (0, target_len - len(y)))
        else:
            y = y[:target_len]

        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel_db = librosa.power_to_db(mel, ref=np.max)
        mel_db = cv2.resize(mel_db, SPEC_SIZE)
        mel_db = np.expand_dims(mel_db, -1)
        mel_db = (mel_db - np.mean(mel_db)) / (np.std(mel_db) + 1e-9)

        text = ""
        try:
            text = stt_model.transcribe(tmp_wav)['text']
        except:
            text = ""

        safe_remove(tmp_wav)
        return mel_db.astype(np.float32), clean_text(text)

    except:
        safe_remove(tmp_wav)
        return np.zeros((SPEC_SIZE[1], SPEC_SIZE[0], 1), dtype=np.float32), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = np.clip(frame, 0, 255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return clean_text(text)
    except:
        return ""

# ===============================
# BUILD MANIFEST
# ===============================
def build_manifest_from_folders():
    print("Building lightweight manifest from dataset folders...")

    video_paths = []
    labels = []
    texts = []
    specs = []

    if not os.path.exists(BASE_DATASET):
        raise FileNotFoundError(f"Folder dataset tidak ditemukan: {BASE_DATASET}")

    class_dirs = []
    for folder in os.listdir(BASE_DATASET):
        path = os.path.join(BASE_DATASET, folder)
        if os.path.isdir(path):
            class_dirs.append(folder)

    for class_folder in tqdm(class_dirs, desc="Scan folders", ascii=True, dynamic_ncols=True):
        class_path = os.path.join(BASE_DATASET, class_folder)
        norm_folder = normalize_label(class_folder)

        if norm_folder not in CLASS_MAP:
            continue

        label = CLASS_MAP[norm_folder]

        file_list = []
        for root, _, files in os.walk(class_path):
            for f in files:
                if f.lower().endswith(VIDEO_EXTENSIONS):
                    file_list.append(os.path.join(root, f))

        for full_path in tqdm(file_list, desc=f"Extract {class_folder}", ascii=True, dynamic_ncols=True, leave=False):
            try:
                middle_frame = extract_middle_frame(full_path)
                spec, text_audio = extract_audio(full_path)
                text_ocr = extract_text_ocr(middle_frame)
                combined_text = clean_text(f"{text_audio} {text_ocr}")

                video_paths.append(full_path)
                labels.append(label)
                texts.append(combined_text)
                specs.append(spec.astype(np.float16))

            except:
                continue

    if len(video_paths) == 0:
        raise ValueError("Tidak ada video yang berhasil diproses. Periksa struktur folder dataset.")

    video_paths = np.array(video_paths, dtype=str)
    labels = np.array(labels, dtype=np.int32)
    texts = np.array(texts, dtype=str)
    specs = np.array(specs, dtype=np.float16)

    np.savez_compressed(
        MANIFEST_CACHE_FILE,
        video_paths=video_paths,
        labels=labels,
        texts=texts,
        specs=specs,
        cache_version=np.array(CACHE_VERSION)
    )

    print(f"Manifest disimpan ke {MANIFEST_CACHE_FILE}")
    return {
        "video_paths": video_paths,
        "labels": labels,
        "texts": texts,
        "specs": specs
    }

def load_or_build_manifest():
    manifest = try_use_processed_data_file()
    if manifest is not None:
        print(f"Using compatible cache: {PROCESSED_DATA_FILE}")
        return manifest

    manifest = load_manifest_if_compatible(MANIFEST_CACHE_FILE)
    if manifest is not None:
        print(f"Using compatible cache: {MANIFEST_CACHE_FILE}")
        return manifest

    return build_manifest_from_folders()

# ===============================
# FUSION LAYER
# ===============================
class AdaptiveFusion(layers.Layer):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.gate = layers.Dense(3, activation='softmax')

    def call(self, inputs):
        v, a, t = inputs
        combined = tf.concat([v, a, t], axis=-1)
        weights = self.gate(combined)

        wv = tf.expand_dims(weights[:, 0], -1)
        wa = tf.expand_dims(weights[:, 1], -1)
        wt = tf.expand_dims(weights[:, 2], -1)

        v = v * wv
        a = a * wa
        t = t * wt

        return tf.concat([v, a, t], axis=-1)

# ===============================
# DATA PIPELINE
# ===============================
class MultiModalSequence(tf.keras.utils.Sequence):
    def __init__(self, video_paths, labels, texts, specs, batch_size=BATCH_SIZE, shuffle=True):
        self.video_paths = np.array(video_paths)
        self.labels = np.array(labels)
        self.texts = np.array(texts)
        self.specs = np.array(specs)
        self.batch_size = batch_size
        self.shuffle = shuffle
        self.indices = np.arange(len(self.labels))
        self.on_epoch_end()

    def __len__(self):
        return int(np.ceil(len(self.indices) / self.batch_size))

    def on_epoch_end(self):
        if self.shuffle:
            np.random.shuffle(self.indices)

    def __getitem__(self, idx):
        batch_ids = self.indices[idx * self.batch_size:(idx + 1) * self.batch_size]

        batch_videos = []
        batch_specs = []
        batch_texts = []
        batch_y = []

        for i in batch_ids:
            video_path = self.video_paths[i]
            frames = extract_frames(video_path, NUM_FRAMES)
            spec = self.specs[i].astype(np.float32)
            text = str(self.texts[i])
            label = np.int32(self.labels[i])

            batch_videos.append(frames)
            batch_specs.append(spec)
            batch_texts.append(text)
            batch_y.append(label)

        batch_videos = np.array(batch_videos, dtype=np.float32)
        batch_specs = np.array(batch_specs, dtype=np.float32)
        batch_y = np.array(batch_y, dtype=np.int32)
        batch_texts = tf.convert_to_tensor(batch_texts, dtype=tf.string)

        return (batch_videos, batch_specs, batch_texts), batch_y

# ===============================
# MODEL
# ===============================
def build_model(text_vectorizer):
    video_in = layers.Input(shape=(NUM_FRAMES, IMG_SIZE[1], IMG_SIZE[0], 3), name="video_input")
    spec_in = layers.Input(shape=(SPEC_SIZE[1], SPEC_SIZE[0], 1), name="spec_input")
    text_in = layers.Input(shape=(), dtype=tf.string, name="text_input")

    video_aug = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.05),
        layers.RandomZoom(0.05),
        layers.RandomContrast(0.1),
    ], name="video_augmentation")

    x_video = layers.TimeDistributed(video_aug, name="td_augmentation")(video_in)

    frame_input = layers.Input(shape=(IMG_SIZE[1], IMG_SIZE[0], 3))
    backbone = EfficientNetB0(
        include_top=False,
        weights="imagenet",
        input_tensor=frame_input
    )
    backbone.trainable = False

    frame_feat = layers.GlobalAveragePooling2D()(backbone.output)
    frame_encoder = models.Model(frame_input, frame_feat, name="frame_encoder")

    video_seq = layers.TimeDistributed(frame_encoder, name="td_frame_encoder")(x_video)
    v = layers.Bidirectional(layers.GRU(128, dropout=0.3, recurrent_dropout=0.1))(video_seq)
    v = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(v)
    v = layers.BatchNormalization()(v)
    v = layers.Dropout(0.3)(v)

    # Audio branch
    a = layers.Conv2D(32, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(spec_in)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Dropout(0.2)(a)

    a = layers.Conv2D(64, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Dropout(0.2)(a)

    a = layers.Conv2D(64, 3, activation='relu', padding='same',
                      kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.BatchNormalization()(a)
    a = layers.MaxPooling2D()(a)

    # (16, 16, 64) -> (256, 64)
    a = layers.Reshape((256, 64))(a)
    a = layers.Bidirectional(layers.GRU(64, dropout=0.2, recurrent_dropout=0.1))(a)
    a = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.Dropout(0.3)(a)

    # Text branch
    t = text_vectorizer(text_in)
    t = layers.Embedding(MAX_VOCAB, 128, mask_zero=True)(t)
    t = layers.Bidirectional(layers.GRU(64, dropout=0.2, recurrent_dropout=0.1))(t)
    t = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(t)
    t = layers.Dropout(0.3)(t)

    fused = AdaptiveFusion()([v, a, t])
    fused = layers.LayerNormalization()(fused)

    x = layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    x = layers.Dropout(0.3)(x)
    out = layers.Dense(1, activation='sigmoid', dtype='float32')(x)

    model = models.Model([video_in, spec_in, text_in], out)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    return model, backbone

def unfreeze_backbone(backbone_model, fine_tune_at=300):
    backbone_model.trainable = True

    for layer in backbone_model.layers[:fine_tune_at]:
        layer.trainable = False

    for layer in backbone_model.layers[fine_tune_at:]:
        if not isinstance(layer, layers.BatchNormalization):
            layer.trainable = True

# ===============================
# PROGRESS CALLBACK
# ===============================
class TQDMTrainingCallback(tf.keras.callbacks.Callback):
    def __init__(self, total_epochs, train_steps, val_steps, label="Training"):
        super().__init__()
        self.total_epochs = int(total_epochs)
        self.train_steps = int(train_steps)
        self.val_steps = int(val_steps)
        self.label = label
        self.epoch_bar = None
        self.batch_bar = None
        self.val_bar = None

    def on_train_begin(self, logs=None):
        self.epoch_bar = tqdm(total=self.total_epochs, desc=self.label, ascii=True, dynamic_ncols=True)

    def on_epoch_begin(self, epoch, logs=None):
        self.batch_bar = tqdm(total=self.train_steps, desc=f"Epoch {epoch+1}/{self.total_epochs}", ascii=True, dynamic_ncols=True, leave=False)

    def on_train_batch_end(self, batch, logs=None):
        if self.batch_bar is not None:
            self.batch_bar.update(1)
            if logs:
                post = {}
                if "loss" in logs:
                    post["loss"] = format_float(logs["loss"])
                if "accuracy" in logs:
                    post["acc"] = format_float(logs["accuracy"])
                self.batch_bar.set_postfix(post, refresh=False)

    def on_test_begin(self, logs=None):
        if self.val_steps > 0:
            self.val_bar = tqdm(total=self.val_steps, desc="Val", ascii=True, dynamic_ncols=True, leave=False)

    def on_test_batch_end(self, batch, logs=None):
        if self.val_bar is not None:
            self.val_bar.update(1)
            if logs:
                post = {}
                if "loss" in logs:
                    post["val_loss"] = format_float(logs["loss"])
                if "accuracy" in logs:
                    post["val_acc"] = format_float(logs["accuracy"])
                self.val_bar.set_postfix(post, refresh=False)

    def on_epoch_end(self, epoch, logs=None):
        if self.batch_bar is not None:
            self.batch_bar.close()
            self.batch_bar = None
        if self.val_bar is not None:
            self.val_bar.close()
            self.val_bar = None

        if logs:
            post = {}
            for key in ["loss", "accuracy", "val_loss", "val_accuracy", "auc", "val_auc", "precision", "val_precision", "recall", "val_recall"]:
                if key in logs and logs[key] is not None:
                    post[key] = format_float(logs[key])
            self.epoch_bar.set_postfix(post, refresh=False)

        self.epoch_bar.update(1)

    def on_train_end(self, logs=None):
        if self.epoch_bar is not None:
            self.epoch_bar.close()
            self.epoch_bar = None

# ===============================
# VISUALIZATION
# ===============================
def visualize_model_architecture(model):
    from tensorflow.keras.utils import plot_model
    plot_model(model, to_file="model_architecture.png", show_shapes=True, show_layer_names=True)
    print("Model architecture saved as model_architecture.png")

def plot_training_history(history_dict):
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    plt.plot(history_dict.get('accuracy', []), label='Train Accuracy')
    plt.plot(history_dict.get('val_accuracy', []), label='Val Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history_dict.get('loss', []), label='Train Loss')
    plt.plot(history_dict.get('val_loss', []), label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap='Blues',
        xticklabels=['Non Kasar', 'Kasar'],
        yticklabels=['Non Kasar', 'Kasar']
    )
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.show()

def merge_histories(h1, h2):
    if h1 is None:
        return h2.copy()
    if h2 is None:
        return h1.copy()

    merged = {}
    keys = set(h1.keys()).union(h2.keys())
    for k in keys:
        v1 = h1.get(k, [])
        v2 = h2.get(k, [])
        merged[k] = list(v1) + list(v2)
    return merged

# ===============================
# PREDICTION
# ===============================
def predict_file(file_path, model):
    frames = extract_frames(file_path, NUM_FRAMES)
    spec, text_audio = extract_audio(file_path)
    middle_frame = frames[NUM_FRAMES // 2]
    text_ocr = extract_text_ocr(middle_frame)

    combined_text = clean_text(f"{text_audio} {text_ocr}")

    video_input = np.expand_dims(frames, axis=0).astype(np.float32)
    spec_input = np.expand_dims(spec, axis=0).astype(np.float32)
    text_input = np.array([combined_text], dtype=object)

    pred = model.predict([video_input, spec_input, text_input], verbose=0)
    prob_kasar = float(pred[0][0])

    pred_class = 1 if prob_kasar >= 0.5 else 0
    category = "Kasar" if pred_class == 1 else "Non Kasar"

    return category, prob_kasar

def get_video_files_recursive(folder_path):
    video_files = []
    for root, _, files in os.walk(folder_path):
        for f in files:
            if f.lower().endswith(VIDEO_EXTENSIONS):
                video_files.append(os.path.join(root, f))
    return video_files

def predict_folder(folder_path, model, num_files=10):
    results = {}

    if not os.path.exists(folder_path):
        return results

    files = get_video_files_recursive(folder_path)
    if len(files) == 0:
        return results

    chosen_files = random.sample(files, num_files) if len(files) > num_files else files

    for file_path in chosen_files:
        file_name = os.path.basename(file_path)
        try:
            category, confidence = predict_file(file_path, model)
            results[file_name] = {
                "file_path": file_path,
                "category": category,
                "confidence": confidence
            }
        except Exception as e:
            results[file_name] = {
                "file_path": file_path,
                "category": "Error",
                "confidence": None,
                "error": str(e)
            }

    return results

# ===============================
# TRAINING
# ===============================
def main_train():
    manifest = load_or_build_manifest()

    video_paths = manifest["video_paths"]
    labels = manifest["labels"]
    texts = manifest["texts"]
    specs = manifest["specs"]

    if len(np.unique(labels)) < 2:
        raise ValueError("Dataset masih hanya berisi satu kelas. Pastikan folder Kasar dan Nonkasar keduanya berisi video.")

    idx = np.arange(len(labels))
    train_idx, val_idx = train_test_split(
        idx,
        test_size=0.2,
        stratify=labels,
        random_state=SEED
    )

    video_paths_train = video_paths[train_idx]
    y_train = labels[train_idx]
    texts_train = texts[train_idx]
    specs_train = specs[train_idx]

    video_paths_val = video_paths[val_idx]
    y_val = labels[val_idx]
    texts_val = texts[val_idx]
    specs_val = specs[val_idx]

    # class_weight tetap dipakai
    weights = compute_class_weight(
        class_weight='balanced',
        classes=np.unique(y_train),
        y=y_train
    )
    class_weight = dict(zip(np.unique(y_train), weights))
    print("Class weight:", class_weight)

    text_vectorizer = layers.TextVectorization(
        max_tokens=MAX_VOCAB,
        output_mode='int',
        output_sequence_length=MAX_TEXT_LEN,
        standardize='lower_and_strip_punctuation',
        name='text_vectorizer'
    )
    text_vectorizer.adapt(texts_train)

    train_seq = MultiModalSequence(
        video_paths_train, y_train, texts_train, specs_train,
        batch_size=BATCH_SIZE, shuffle=True
    )
    val_seq = MultiModalSequence(
        video_paths_val, y_val, texts_val, specs_val,
        batch_size=BATCH_SIZE, shuffle=False
    )

    if os.path.exists(MODEL_FILE):
        try:
            print(f"Model ditemukan di {MODEL_FILE}, memuat model...")
            model = tf.keras.models.load_model(
                MODEL_FILE,
                custom_objects={'AdaptiveFusion': AdaptiveFusion}
            )

            preds = model.predict(val_seq, verbose=0)
            preds_class = (preds > 0.5).astype(int).flatten()

            print(classification_report(
                y_val,
                preds_class,
                target_names=['Non Kasar', 'Kasar'],
                digits=4
            ))
            plot_confusion_matrix(y_val, preds_class)
            return

        except Exception as e:
            print(f"Model lama gagal dimuat, training ulang. Alasan: {e}")

    model, backbone = build_model(text_vectorizer)

    callbacks_stage1 = [
        TQDMTrainingCallback(
            total_epochs=EPOCHS_STAGE1,
            train_steps=len(train_seq),
            val_steps=len(val_seq),
            label="Stage1"
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=5,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=2,
            min_lr=1e-6
        ),
        tf.keras.callbacks.ModelCheckpoint(
            MODEL_FILE,
            monitor='val_loss',
            save_best_only=True
        ),
        tf.keras.callbacks.TerminateOnNaN()
    ]

    history1 = model.fit(
        train_seq,
        validation_data=val_seq,
        epochs=EPOCHS_STAGE1,
        callbacks=callbacks_stage1,
        class_weight=class_weight,
        verbose=0
    )

    unfreeze_backbone(backbone, fine_tune_at=max(0, len(backbone.layers) - 30))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-5),
        loss='binary_crossentropy',
        metrics=[
            'accuracy',
            tf.keras.metrics.AUC(name='auc'),
            tf.keras.metrics.Precision(name='precision'),
            tf.keras.metrics.Recall(name='recall')
        ]
    )

    callbacks_stage2 = [
        TQDMTrainingCallback(
            total_epochs=EPOCHS_STAGE2,
            train_steps=len(train_seq),
            val_steps=len(val_seq),
            label="Stage2"
        ),
        tf.keras.callbacks.EarlyStopping(
            monitor='val_loss',
            patience=4,
            restore_best_weights=True
        ),
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=2,
            min_lr=1e-7
        ),
        tf.keras.callbacks.ModelCheckpoint(
            MODEL_FILE,
            monitor='val_loss',
            save_best_only=True
        ),
        tf.keras.callbacks.TerminateOnNaN()
    ]

    history2 = model.fit(
        train_seq,
        validation_data=val_seq,
        epochs=EPOCHS_STAGE2,
        callbacks=callbacks_stage2,
        class_weight=class_weight,
        verbose=0
    )

    history_all = merge_histories(history1.history, history2.history)
    np.savez(
        HISTORY_FILE,
        **{k: np.array(v) for k, v in history_all.items()}
    )
    print(f"Training history saved to {HISTORY_FILE}")

    visualize_model_architecture(model)

    preds = model.predict(val_seq, verbose=0)
    preds_class = (preds > 0.5).astype(int).flatten()

    print(classification_report(
        y_val,
        preds_class,
        target_names=['Non Kasar', 'Kasar'],
        digits=4
    ))

    plot_confusion_matrix(y_val, preds_class)
    plot_training_history(history_all)

# ===============================
# PREDICTION MODE
# ===============================
def main_predict():
    if not os.path.exists(MODEL_FILE):
        print("Model belum ditemukan, pastikan model sudah dilatih dan disimpan.")
        return

    model = tf.keras.models.load_model(
        MODEL_FILE,
        custom_objects={'AdaptiveFusion': AdaptiveFusion}
    )

    folders = ["Kasar", "Nonkasar"]

    for folder in folders:
        folder_path = os.path.join(BASE_DATASET, folder)
        results = predict_folder(folder_path, model, num_files=10)

        print(f"\nHasil Prediksi untuk Folder '{folder}':")
        for file_name, result in results.items():
            print(f"File: {result['file_path']}")
            print(f"  Kategori: {result['category']}")
            if result.get("confidence") is not None:
                print(f"  Confidence Kasar: {result['confidence']:.4f}")
            if "error" in result:
                print(f"  Error: {result['error']}")

# ===============================
# RUN
# ===============================
if __name__ == "__main__":
    main_train()
    main_predict()
```
