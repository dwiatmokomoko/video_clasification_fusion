```python
# -*- coding: utf-8 -*-
"""
Deteksi kata kasar pada video pendek (satu file utuh).
Perbaikan:
 - Transformer block: feed-forward menghasilkan dimensi yang sama sebelum Add
 - Reshape audio menjadi (1024,128) sesuai dua pooling dari SPEC_SIZE 128x128
 - Tetap menggunakan cache fitur (generated_dataset/*.npz)
 - Augmentasi, plotting, confusion matrix, realtime inference
"""
import os
import cv2
import re
import random
import numpy as np
import tensorflow as tf
import librosa
import whisper
import easyocr
from moviepy import VideoFileClip
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
from datetime import datetime
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import itertools
import warnings
warnings.filterwarnings("ignore")

# =====================================
# CONFIG
# =====================================

BASE_DATASET = "dataset"            # struktur: dataset/aman, dataset/anjing, ...
GENERATED_DATASET = "generated_dataset"  # cache fitur per video

IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100

SR = 22050
DURATION = 2  # detik audio yang diambil

BATCH_SIZE = 8
EPOCHS = 25
VALIDATION_SPLIT = 0.2
RANDOM_SEED = 42

os.makedirs(GENERATED_DATASET, exist_ok=True)
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
tf.random.set_seed(RANDOM_SEED)

# =====================================
# LOAD OPTIONAL MODELS (Whisper & OCR)
# =====================================

print("Loading Whisper & OCR (optional, may take time)...")
try:
    stt_model = whisper.load_model("base")
except Exception as e:
    print("Warning: Whisper gagal dimuat:", e)
    stt_model = None

try:
    ocr_reader = easyocr.Reader(['id','en'], gpu=False)
except Exception as e:
    print("Warning: EasyOCR gagal dimuat:", e)
    ocr_reader = None

# =====================================
# UTIL: plotting
# =====================================

def plot_history(history, out_dir="plots"):
    os.makedirs(out_dir, exist_ok=True)
    plt.figure(figsize=(12,5))
    # loss
    plt.subplot(1,2,1)
    plt.plot(history.history['loss'], label='train_loss')
    if 'val_loss' in history.history:
        plt.plot(history.history['val_loss'], label='val_loss')
    plt.title('Loss')
    plt.legend()
    # acc
    plt.subplot(1,2,2)
    acc_key = 'accuracy' if 'accuracy' in history.history else 'acc'
    plt.plot(history.history[acc_key], label='train_acc')
    if 'val_' + acc_key in history.history:
        plt.plot(history.history['val_' + acc_key], label='val_acc')
    plt.title('Accuracy')
    plt.legend()
    fname = os.path.join(out_dir, "training_history.png")
    plt.savefig(fname)
    plt.close()
    print("Saved training plot to", fname)

def plot_confusion_matrix(cm, classes, out_path, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):
    plt.figure(figsize=(6,6))
    if normalize:
        cm = cm.astype('float') / (cm.sum(axis=1)[:, np.newaxis] + 1e-9)
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
    plt.savefig(out_path)
    plt.close()
    print("Saved confusion matrix to", out_path)

# =====================================
# FEATURE EXTRACTION (with caching)
# =====================================

def _safe_filename(s):
    return re.sub(r'[^0-9a-zA-Z_\-\.]', '_', s)

def extract_audio(video_path):
    tmp = os.path.join(GENERATED_DATASET, "tmp_audio.wav")
    with VideoFileClip(video_path) as video:
        if video.audio is None:
            raise ValueError("Video tidak memiliki audio")
        video.audio.write_audiofile(tmp, logger=None)
    y, sr = librosa.load(tmp, sr=SR)
    target = int(SR * DURATION)
    if len(y) < target:
        y = np.pad(y, (0, target - len(y)))
    y = y[:target]
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
    mel = librosa.power_to_db(mel)
    mel = cv2.resize(mel, SPEC_SIZE)
    mel = np.expand_dims(mel, -1).astype(np.float32)
    try:
        os.remove(tmp)
    except:
        pass
    return mel

def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT)) or 0
    if total > 0:
        cap.set(cv2.CAP_PROP_POS_FRAMES, total // 2)
    ret, frame = cap.read()
    cap.release()
    if not ret or frame is None:
        return np.zeros((*IMG_SIZE, 3), dtype=np.float32)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    frame = cv2.resize(frame, IMG_SIZE)
    frame = frame / 255.0
    return frame.astype(np.float32)

def extract_text(video_path, frame):
    tmp = os.path.join(GENERATED_DATASET, "tmp_audio.wav")
    text_parts = []
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is not None:
                video.audio.write_audiofile(tmp, logger=None)
                if stt_model is not None:
                    stt = stt_model.transcribe(tmp)['text']
                    text_parts.append(stt)
    except Exception as e:
        print("Warning STT failed:", e)
    try:
        if ocr_reader is not None:
            if frame.max() <= 1:
                frame_uint8 = (frame * 255).astype(np.uint8)
            else:
                frame_uint8 = frame.astype(np.uint8)
            ocr = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
            text_parts.append(ocr)
    except Exception as e:
        print("Warning OCR failed:", e)
    try:
        os.remove(tmp)
    except:
        pass
    text = " ".join(text_parts).lower()
    text = re.sub(r'(.)\1+', r'\1', text)
    seq = [ord(c) % 5000 for c in text][:MAX_TEXT_LEN]
    seq = seq + [0] * (MAX_TEXT_LEN - len(seq))
    return np.array(seq, dtype=np.int32)

def cache_path_for_video(video_path):
    rel = os.path.relpath(video_path)
    fname = _safe_filename(rel) + ".npz"
    return os.path.join(GENERATED_DATASET, fname)

def process_and_cache(video_path, label):
    cache_path = cache_path_for_video(video_path)
    if os.path.exists(cache_path):
        try:
            data = np.load(cache_path)
            img = data['img']
            spec = data['spec']
            txt = data['txt']
            lbl = int(data['label'])
            return img, spec, txt, lbl
        except Exception:
            try:
                os.remove(cache_path)
            except:
                pass
    img = extract_frame(video_path)
    spec = extract_audio(video_path)
    txt = extract_text(video_path, img)
    try:
        np.savez_compressed(cache_path, img=img, spec=spec, txt=txt, label=int(label))
    except Exception as e:
        print("Warning: gagal menyimpan cache:", e)
    return img, spec, txt, int(label)

# =====================================
# DATA AUGMENTATION (numpy)
# =====================================

def augment_sample_np(img, spec, txt):
    # 1) Random horizontal flip
    if random.random() < 0.5:
        img = np.fliplr(img).copy()
    # 2) Audio time masking (mask along time axis)
    spec2 = spec.copy()
    time_dim = spec2.shape[1]
    num_masks = random.randint(0, 2)
    for _ in range(num_masks):
        t = random.randint(1, max(1, time_dim // 8))
        t0 = random.randint(0, max(0, time_dim - t))
        spec2[:, t0:t0+t, 0] = spec2.min()
    # 3) Random deletion for text tokens
    txt2 = txt.copy()
    delete_prob = 0.12
    for i in range(len(txt2)):
        if txt2[i] != 0 and random.random() < delete_prob:
            txt2[i] = 0
    return img.astype(np.float32), spec2.astype(np.float32), txt2.astype(np.int32)

# IMPORTANT: tf_augment must accept ((img,spec,txt), label)
def tf_augment(inputs, label):
    img, spec, txt = inputs[0], inputs[1], inputs[2]
    def _aug(i, s, t):
        ai, aspec, at = augment_sample_np(i, s, t)
        return ai, aspec, at
    img_aug, spec_aug, txt_aug = tf.numpy_function(_aug, [img, spec, txt], [tf.float32, tf.float32, tf.int32])
    img_aug.set_shape((IMG_SIZE[0], IMG_SIZE[1], 3))
    spec_aug.set_shape((SPEC_SIZE[0], SPEC_SIZE[1], 1))
    txt_aug.set_shape((MAX_TEXT_LEN,))
    return (img_aug, spec_aug, txt_aug), label

# =====================================
# DATASET LOADER (memanfaatkan cache)
# =====================================

def generate_dataset_from_cache():
    imgs = []
    specs = []
    texts = []
    labels = []
    class_names = []
    for folder in sorted(os.listdir(BASE_DATASET)):
        folder_path = os.path.join(BASE_DATASET, folder)
        if not os.path.isdir(folder_path):
            continue
        class_names.append(folder)
        label = 0 if folder.lower() == "aman" else 1
        for file in sorted(os.listdir(folder_path)):
            if not file.lower().endswith((".mp4", ".mkv", ".avi", ".mov")):
                continue
            path = os.path.join(folder_path, file)
            try:
                img, spec, txt, lbl = process_and_cache(path, label)
                imgs.append(img)
                specs.append(spec)
                texts.append(txt)
                labels.append(lbl)
            except Exception as e:
                print("❌ Skipping:", path, "Reason:", e)
                continue
    if len(imgs) == 0:
        raise RuntimeError("Tidak ada data yang berhasil dimuat. Periksa folder dataset.")
    X_img = np.array(imgs, dtype=np.float32)
    X_spec = np.array(specs, dtype=np.float32)
    X_txt = np.array(texts, dtype=np.int32)
    y = np.array(labels, dtype=np.int32)
    print("Total samples (cached):", len(y))
    return X_img, X_spec, X_txt, y, class_names

# =====================================
# MODEL BUILD
# =====================================

def transformer_block(x):
    # MultiHeadAttention + residual + FFN (FFN menghasilkan dimensi sama dengan x)
    attn = layers.MultiHeadAttention(num_heads=4, key_dim=64)(x, x)
    x = layers.Add()([x, attn])
    x = layers.LayerNormalization()(x)
    d_model = int(x.shape[-1]) if x.shape[-1] is not None else None
    if d_model is None:
        # fallback: gunakan 64 jika tidak diketahui (jarang terjadi)
        d_model = 64
    ff = layers.Dense(max(64, d_model), activation='relu')(x)
    ff = layers.Dense(d_model)(ff)
    x = layers.Add()([x, ff])
    x = layers.LayerNormalization()(x)
    return x

def build_model():
    img_in = layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3), name="img_in")
    spec_in = layers.Input(shape=(SPEC_SIZE[0], SPEC_SIZE[1], 1), name="spec_in")
    txt_in = layers.Input(shape=(MAX_TEXT_LEN,), name="txt_in")

    # Visual branch: EfficientNetB0 (fallback jika tidak tersedia)
    try:
        base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
        v = layers.GlobalAveragePooling2D()(base.output)
    except Exception as e:
        print("EfficientNetB0 tidak tersedia, menggunakan Conv fallback. Error:", e)
        x = layers.Conv2D(32, 3, activation='relu', padding='same')(img_in)
        x = layers.MaxPooling2D()(x)
        x = layers.Conv2D(64, 3, activation='relu', padding='same')(x)
        v = layers.GlobalAveragePooling2D()(x)

    # Audio branch: CNN -> BiGRU
    a = layers.Conv2D(64, (3,3), activation='relu', padding='same')(spec_in)
    a = layers.MaxPooling2D((2,2))(a)   # 128 -> 64
    a = layers.Conv2D(128, (3,3), activation='relu', padding='same')(a)
    a = layers.MaxPooling2D((2,2))(a)   # 64 -> 32
    # After two poolings, spatial dims = 32 x 32 (SPEC_SIZE 128 -> 32)
    # Reshape to sequence: (32*32, channels=128) => (1024,128)
    a_seq = layers.Reshape(( (SPEC_SIZE[0]//4) * (SPEC_SIZE[1]//4), 128 ))(a)  # (1024,128)
    a_seq = layers.Bidirectional(layers.GRU(128, return_sequences=True))(a_seq)  # output dim = 256
    a_seq = transformer_block(a_seq)
    a = layers.GlobalAveragePooling1D()(a_seq)

    # Text branch: Embedding + BiLSTM + Transformer-like
    t = layers.Embedding(input_dim=5000, output_dim=128, mask_zero=True)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64, return_sequences=True))(t)
    t = transformer_block(t)
    t = layers.GlobalAveragePooling1D()(t)

    # Fusion
    x = layers.concatenate([v, a, t])
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128, activation='relu')(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model(inputs=[img_in, spec_in, txt_in], outputs=out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4), loss='binary_crossentropy', metrics=['accuracy'])
    model.summary()
    return model

# =====================================
# TRAINING PIPELINE (memakai cache)
# =====================================

def train_and_evaluate():
    X_img, X_spec, X_txt, y, class_names = generate_dataset_from_cache()

    # shuffle & split
    idx = np.arange(len(y))
    np.random.shuffle(idx)
    X_img = X_img[idx]
    X_spec = X_spec[idx]
    X_txt = X_txt[idx]
    y = y[idx]

    split = int(len(y) * (1 - VALIDATION_SPLIT))
    X_img_train, X_img_val = X_img[:split], X_img[split:]
    X_spec_train, X_spec_val = X_spec[:split], X_spec[split:]
    X_txt_train, X_txt_val = X_txt[:split], X_txt[split:]
    y_train, y_val = y[:split], y[split:]

    print("Train samples:", len(y_train), "Val samples:", len(y_val))

    train_ds = tf.data.Dataset.from_tensor_slices(((X_img_train, X_spec_train, X_txt_train), y_train))
    val_ds = tf.data.Dataset.from_tensor_slices(((X_img_val, X_spec_val, X_txt_val), y_val))

    # IMPORTANT: augment expects ((img,spec,txt), label)
    train_ds = train_ds.shuffle(500).map(tf_augment, num_parallel_calls=tf.data.AUTOTUNE)
    train_ds = train_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    val_ds = val_ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    # strategy
    try:
        strategy = tf.distribute.MirroredStrategy()
        print("Using MirroredStrategy, devices:", strategy.num_replicas_in_sync)
    except Exception:
        strategy = None

    with (strategy.scope() if strategy is not None else tf.device('/CPU:0')):
        model = build_model()

    log_dir = "logs/" + datetime.now().strftime("%Y%m%d-%H%M%S")
    callbacks = [
        tf.keras.callbacks.ModelCheckpoint("best_model.h5", save_best_only=True, monitor='val_loss'),
        tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True),
        tf.keras.callbacks.TensorBoard(log_dir=log_dir)
    ]

    history = model.fit(train_ds, validation_data=val_ds, epochs=EPOCHS, callbacks=callbacks)

    model.save("final_hybrid_model.h5")
    print("Model saved to final_hybrid_model.h5")

    plot_history(history)

    # EVALUASI pada val set
    preds = model.predict([X_img_val, X_spec_val, X_txt_val], batch_size=BATCH_SIZE)
    y_pred = (preds.flatten() > 0.5).astype(int)
    cm = confusion_matrix(y_val, y_pred)
    report = classification_report(y_val, y_pred, target_names=["AMAN", "KASAR"], digits=4)
    print("Classification Report:\n", report)
    os.makedirs("plots", exist_ok=True)
    plot_confusion_matrix(cm, classes=["AMAN", "KASAR"], out_path=os.path.join("plots", "confusion_matrix.png"), normalize=False)
    plot_confusion_matrix(cm, classes=["AMAN", "KASAR"], out_path=os.path.join("plots", "confusion_matrix_norm.png"), normalize=True)

    # simpan per-sample hasil validasi
    results_path = os.path.join("plots", "val_results.txt")
    with open(results_path, "w", encoding="utf-8") as f:
        f.write("idx\ttrue\tpred\tprob\n")
        for i, (t, p) in enumerate(zip(y_val, preds.flatten())):
            f.write(f"{i}\t{t}\t{int(p>0.5)}\t{p:.4f}\n")
    print("Saved per-sample validation results to", results_path)

    return model, (X_img_val, X_spec_val, X_txt_val, y_val, y_pred, preds), class_names

# =====================================
# REALTIME DETECTION (kamera)
# =====================================

def realtime(model_path="final_hybrid_model.h5"):
    if not os.path.exists(model_path):
        print("Model tidak ditemukan:", model_path)
        return
    model = tf.keras.models.load_model(model_path)
    cap = cv2.VideoCapture(0)
    print("Tekan ESC untuk keluar.")
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = cv2.resize(frame_rgb, IMG_SIZE) / 255.0
        img = np.expand_dims(img.astype(np.float32), 0)
        spec = np.zeros((1, SPEC_SIZE[0], SPEC_SIZE[1], 1), dtype=np.float32)
        txt = np.zeros((1, MAX_TEXT_LEN), dtype=np.int32)
        pred = model.predict([img, spec, txt])[0][0]
        label = "KASAR" if pred > 0.5 else "AMAN"
        color = (0,0,255) if pred > 0.5 else (0,255,0)
        cv2.putText(frame, f"{label} {pred:.2f}", (30,50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2)
        cv2.imshow("Realtime Detection", frame)
        if cv2.waitKey(1) == 27:
            break
    cap.release()
    cv2.destroyAllWindows()

# =====================================
# MAIN
# =====================================

if __name__ == "__main__":
    model, val_info, class_names = train_and_evaluate()
    X_img_v, X_spec_v, X_txt_v, y_true, y_pred, preds = val_info
    print("Contoh prediksi (first 10):")
    for i in range(min(10, len(y_true))):
        print(f"idx={i} true={y_true[i]} pred={y_pred[i]} prob={preds.flatten()[i]:.4f}")
    # Uncomment untuk realtime:
    # realtime("final_hybrid_model.h5")
```

```python
# Setelah training
model, val_info, class_names = train_and_evaluate()
X_img_v, X_spec_v, X_txt_v, y_true, y_pred, preds = val_info

# Menampilkan plot loss dan accuracy
history_plot_dir = "plots"
os.makedirs(history_plot_dir, exist_ok=True)
plot_history(history)  # ini sudah men-generate plots/training_history.png

# Menampilkan confusion matrix (tanpa normalisasi)
cm = confusion_matrix(y_true, y_pred)
plot_confusion_matrix(cm, classes=class_names,
                      out_path=os.path.join(history_plot_dir, "confusion_matrix.png"),
                      normalize=False,
                      title="Confusion Matrix")

# Menampilkan confusion matrix (dengan normalisasi)
plot_confusion_matrix(cm, classes=class_names,
                      out_path=os.path.join(history_plot_dir, "confusion_matrix_norm.png"),
                      normalize=True,
                      title="Normalized Confusion Matrix")

# Opsional: menampilkan di layar (bukan menyimpan)
plt.figure(figsize=(12,5))

# Loss
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss')
plt.legend()

# Accuracy
plt.subplot(1,2,2)
acc_key = 'accuracy' if 'accuracy' in history.history else 'acc'
plt.plot(history.history[acc_key], label='train_acc')
plt.plot(history.history['val_' + acc_key], label='val_acc')
plt.title('Accuracy')
plt.legend()

plt.show()
```
