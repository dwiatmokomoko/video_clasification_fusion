```python
# -*- coding: utf-8 -*-
"""
FINAL ALL-IN-ONE
Adaptive Multimodal Fusion + Visualization
"""

import os
import cv2
import re
import numpy as np
import tensorflow as tf
import librosa
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils.class_weight import compute_class_weight
import itertools
import warnings
warnings.filterwarnings("ignore")

# =====================================
# CONFIG
# =====================================

BASE_DATASET = "dataset"
IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100

SR = 22050
DURATION = 2

BATCH_SIZE = 8
EPOCHS = 15

# =====================================
# CLEAN FILE NAMES
# =====================================

def clean_filenames():
    for root, _, files in os.walk(BASE_DATASET):
        for f in files:
            old = os.path.join(root, f)
            new_name = re.sub(r'[^a-zA-Z0-9_.-]', '_', f)
            new = os.path.join(root, new_name)
            if old != new:
                try:
                    os.rename(old, new)
                except:
                    pass

# =====================================
# FEATURE EXTRACTION
# =====================================

def extract_frame(video_path):
    try:
        cap = cv2.VideoCapture(video_path)
        total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        cap.set(cv2.CAP_PROP_POS_FRAMES, total//2)
        ret, frame = cap.read()
        cap.release()

        if not ret:
            return np.zeros((*IMG_SIZE,3))

        frame = cv2.resize(frame, IMG_SIZE)
        return (frame/255.0).astype(np.float32)
    except:
        return np.zeros((*IMG_SIZE,3))

def extract_audio(video_path):
    try:
        tmp = "temp.wav"
        with VideoFileClip(video_path) as video:
            if video.audio is None:
                return np.zeros((*SPEC_SIZE,1))
            video.audio.write_audiofile(tmp, logger=None)

        y, sr = librosa.load(tmp, sr=SR)
        y = y[:SR*DURATION]

        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
        mfcc = cv2.resize(mfcc, SPEC_SIZE)
        mfcc = np.expand_dims(mfcc, -1)

        os.remove(tmp)
        return mfcc.astype(np.float32)
    except:
        return np.zeros((*SPEC_SIZE,1))

def extract_text(video_path):
    text = os.path.basename(video_path).lower()
    text = re.sub(r'[^a-z]', '', text)

    seq = [ord(c)%5000 for c in text][:MAX_TEXT_LEN]
    seq += [0]*(MAX_TEXT_LEN-len(seq))

    return np.array(seq)

# =====================================
# LOAD DATA
# =====================================

def load_data():
    imgs, specs, texts, labels = [],[],[],[]

    for folder in os.listdir(BASE_DATASET):
        path = os.path.join(BASE_DATASET, folder)
        if not os.path.isdir(path): continue

        label = 0 if folder.lower()=="aman" else 1

        for f in os.listdir(path):
            if not f.endswith(".mp4"): continue

            full = os.path.join(path, f)

            try:
                img = extract_frame(full)
                spec = extract_audio(full)
                txt = extract_text(full)

                imgs.append(img)
                specs.append(spec)
                texts.append(txt)
                labels.append(label)

            except:
                continue

    print("Total data:", len(labels))
    return np.array(imgs), np.array(specs), np.array(texts), np.array(labels)

# =====================================
# ADAPTIVE FUSION
# =====================================

def adaptive_fusion(v,a,t):
    combined = layers.concatenate([v,a,t])
    weights = layers.Dense(3, activation='softmax')(combined)

    wv = layers.Lambda(lambda x: x[:,0:1])(weights)
    wa = layers.Lambda(lambda x: x[:,1:2])(weights)
    wt = layers.Lambda(lambda x: x[:,2:3])(weights)

    v = layers.Multiply()([v,wv])
    a = layers.Multiply()([a,wa])
    t = layers.Multiply()([t,wt])

    return layers.concatenate([v,a,t])

# =====================================
# MODEL
# =====================================

def build_model():
    img_in = layers.Input(shape=(224,224,3))
    spec_in = layers.Input(shape=(128,128,1))
    txt_in = layers.Input(shape=(100,))

    # VISUAL
    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    base.trainable = False
    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(128, activation='relu')(v)

    # AUDIO
    a = layers.Conv2D(32,3,activation='relu',padding='same')(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64,3,activation='relu',padding='same')(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((1024,64))(a)
    a = layers.Bidirectional(layers.GRU(64))(a)
    a = layers.Dense(128,activation='relu')(a)

    # TEXT
    t = layers.Embedding(5000,128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64))(t)
    t = layers.Dense(128,activation='relu')(t)

    # FUSION
    fused = adaptive_fusion(v,a,t)

    # CLASSIFIER
    x = layers.Dense(256,activation='relu')(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128,activation='relu')(x)

    out = layers.Dense(1,activation='sigmoid')(x)

    model = models.Model([img_in,spec_in,txt_in], out)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# =====================================
# VISUALIZATION
# =====================================

def show_all(history, y_true, preds, X_spec):
    y_pred = (preds > 0.5).astype(int).flatten()

    plt.figure(figsize=(18,5))

    # Accuracy
    plt.subplot(1,3,1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Accuracy History')
    plt.legend(['Train','Val'])

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.subplot(1,3,2)
    plt.imshow(cm, cmap='Blues')
    plt.title('Confusion Matrix')

    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j], ha="center")

    # Spectrogram
    plt.subplot(1,3,3)
    spec = X_spec[0][:,:,0]
    plt.imshow(spec, aspect='auto', cmap='magma')
    plt.title('Sample Mel-Spectrogram')

    plt.tight_layout()
    plt.show()

# =====================================
# MAIN
# =====================================

def main():
    clean_filenames()

    X_img,X_spec,X_txt,y = load_data()

    idx = np.arange(len(y))
    np.random.shuffle(idx)

    X_img,X_spec,X_txt,y = X_img[idx],X_spec[idx],X_txt[idx],y[idx]

    split = int(len(y)*0.8)

    X_img_tr,X_img_val = X_img[:split],X_img[split:]
    X_spec_tr,X_spec_val = X_spec[:split],X_spec[split:]
    X_txt_tr,X_txt_val = X_txt[:split],X_txt[split:]
    y_tr,y_val = y[:split],y[split:]

    cw = compute_class_weight('balanced', classes=np.unique(y_tr), y=y_tr)
    cw = dict(enumerate(cw))

    model = build_model()

    history = model.fit(
        [X_img_tr,X_spec_tr,X_txt_tr],
        y_tr,
        validation_data=([X_img_val,X_spec_val,X_txt_val], y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        class_weight=cw
    )

    preds = model.predict([X_img_val,X_spec_val,X_txt_val])

    print(classification_report(y_val, (preds>0.5).astype(int)))

    show_all(history, y_val, preds, X_spec_val)

# =====================================
# RUN
# =====================================

if __name__ == "__main__":
    main()
```
