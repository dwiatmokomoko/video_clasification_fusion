```python
# -*- coding: utf-8 -*-
"""
FINAL VALID RESEARCH VERSION
Multimodal: Visual + Audio + Text (OCR + Speech-to-Text)
"""

import os
import cv2
import re
import numpy as np
import tensorflow as tf
import librosa
import whisper
import easyocr
import matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models
from tensorflow.keras.applications import EfficientNetB0
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
import itertools
import warnings
warnings.filterwarnings("ignore")

# =====================================
# CONFIG
# =====================================

BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100

SR = 22050
DURATION = 3

BATCH_SIZE = 8
EPOCHS = 15

# =====================================
# LOAD MODELS (OCR + STT)
# =====================================

print("Loading Whisper...")
stt_model = whisper.load_model("base")

print("Loading OCR...")
ocr_reader = easyocr.Reader(['id','en'], gpu=False)

# =====================================
# FEATURE EXTRACTION
# =====================================

def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total//2)
    ret, frame = cap.read()
    cap.release()

    if not ret:
        return np.zeros((*IMG_SIZE,3))

    frame = cv2.resize(frame, IMG_SIZE)
    return (frame/255.0).astype(np.float32)

# -------------------------------------

def extract_audio(video_path):
    try:
        tmp = "temp.wav"

        with VideoFileClip(video_path) as video:
            if video.audio is None:
                return np.zeros((*SPEC_SIZE,1)), ""
            video.audio.write_audiofile(tmp, logger=None)

        y, sr = librosa.load(tmp, sr=SR)

        # padding
        if len(y) < SR*DURATION:
            y = np.pad(y, (0, SR*DURATION-len(y)))

        y = y[:SR*DURATION]

        # normalisasi
        y = y + 0.005*np.random.randn(len(y))

        # spectrogram
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = librosa.power_to_db(mel)
        mel = cv2.resize(mel, SPEC_SIZE)
        mel = np.expand_dims(mel, -1)

        # normalisasi
        mel = (mel - np.mean(mel)) / (np.std(mel) + 1e-9)

        # ===== SPEECH TO TEXT =====
        text = ""
        try:
            result = stt_model.transcribe(tmp)
            text = result['text']
        except:
            pass

        os.remove(tmp)

        return mel.astype(np.float32), text.lower()

    except:
        return np.zeros((*SPEC_SIZE,1)), ""

# -------------------------------------

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame*255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except:
        return ""

# -------------------------------------

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c)%5000 for c in text][:MAX_TEXT_LEN]
    seq += [0]*(MAX_TEXT_LEN-len(seq))
    return np.array(seq)

# =====================================
# LOAD DATA
# =====================================

def load_data():
    imgs, specs, texts, labels = [],[],[],[]

    for folder in os.listdir(BASE_DATASET):
        path = os.path.join(BASE_DATASET, folder)
        if not os.path.isdir(path): continue

        label = 0 if folder.lower()=="aman" else 1

        for f in os.listdir(path):
            if not f.endswith(".mp4"): continue

            full = os.path.join(path, f)

            try:
                frame = extract_frame(full)
                spec, text_audio = extract_audio(full)
                text_ocr = extract_text_ocr(frame)

                # combine text
                full_text = text_audio + " " + text_ocr
                txt_seq = text_to_sequence(full_text)

                imgs.append(frame)
                specs.append(spec)
                texts.append(txt_seq)
                labels.append(label)

            except Exception as e:
                print("Skip:", f, e)
                continue

    print("Total data:", len(labels))
    return np.array(imgs), np.array(specs), np.array(texts), np.array(labels)

# =====================================
# ADAPTIVE FUSION
# =====================================

def adaptive_fusion(v,a,t):
    combined = layers.concatenate([v,a,t])
    weights = layers.Dense(3, activation='softmax')(combined)

    wv = layers.Lambda(lambda x: x[:,0:1])(weights)
    wa = layers.Lambda(lambda x: x[:,1:2])(weights)
    wt = layers.Lambda(lambda x: x[:,2:3])(weights)

    v = layers.Multiply()([v,wv])
    a = layers.Multiply()([a,wa])
    t = layers.Multiply()([t,wt])

    return layers.concatenate([v,a,t])

# =====================================
# MODEL
# =====================================

def build_model():
    img_in = layers.Input(shape=(224,224,3))
    spec_in = layers.Input(shape=(128,128,1))
    txt_in = layers.Input(shape=(100,))

    # VISUAL
    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    base.trainable = False
    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(128, activation='relu')(v)

    # AUDIO
    a = layers.Conv2D(32,3,activation='relu',padding='same')(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64,3,activation='relu',padding='same')(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((1024,64))(a)
    a = layers.Bidirectional(layers.GRU(64))(a)
    a = layers.Dense(128,activation='relu')(a)

    # TEXT
    t = layers.Embedding(5000,128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64))(t)
    t = layers.Dense(128,activation='relu')(t)

    # FUSION
    fused = adaptive_fusion(v,a,t)

    # CLASSIFIER
    x = layers.Dense(256,activation='relu')(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128,activation='relu')(x)

    out = layers.Dense(1,activation='sigmoid')(x)

    model = models.Model([img_in,spec_in,txt_in], out)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(1e-4),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    return model

# =====================================
# VISUALIZATION
# =====================================

def show_all(history, y_true, preds):
    y_pred = (preds > 0.5).astype(int)

    plt.figure(figsize=(15,5))

    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title("Accuracy")

    plt.subplot(1,2,2)
    cm = confusion_matrix(y_true, y_pred)
    plt.imshow(cm, cmap='Blues')
    plt.title("Confusion Matrix")

    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j,i,cm[i,j],ha="center")

    plt.show()

# =====================================
# MAIN
# =====================================

def main():
    X_img,X_spec,X_txt,y = load_data()

    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_txt_tr, X_txt_val, y_tr, y_val = train_test_split(
        X_img, X_spec, X_txt, y, test_size=0.2, random_state=42, stratify=y
    )

    model = build_model()
    model.summary()

    history = model.fit(
        [X_img_tr,X_spec_tr,X_txt_tr],
        y_tr,
        validation_data=([X_img_val,X_spec_val,X_txt_val], y_val),
        epochs=EPOCHS,
        batch_size=BATCH_SIZE
    )

    preds = model.predict([X_img_val,X_spec_val,X_txt_val])

    print(classification_report(y_val, (preds>0.5).astype(int)))

    show_all(history, y_val, preds)

# =====================================
# RUN
# =====================================

if __name__ == "__main__":
    main()
```
