```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings
import tensorflow as tf, librosa, whisper, easyocr, matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import itertools
warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
PROCESSED_DATA_FILE = "processed_data.npz"
MODEL_FILE = "best_model.h5"
HISTORY_FILE = "history.npz"

IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 15

# ===============================
# LOAD MODELS (OCR + STT)
# ===============================
print("Loading Whisper...")
stt_model = whisper.load_model("base")

print("Loading OCR...")
ocr_reader = easyocr.Reader(['id','en'], gpu=False)

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total//2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE,3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame/255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = "temp.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE,1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR*DURATION: y = np.pad(y, (0, SR*DURATION-len(y)))
        y = y[:SR*DURATION] + 0.005*np.random.randn(SR*DURATION)
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel))/(np.std(mel)+1e-9)
        text = ""
        try: text = stt_model.transcribe(tmp)['text']
        except: pass
        os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except:
        return np.zeros((*SPEC_SIZE,1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame*255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except: return ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c)%5000 for c in text][:MAX_TEXT_LEN]
    seq += [0]*(MAX_TEXT_LEN-len(seq))
    return np.array(seq)

# ===============================
# LOAD / CACHE DATA
# ===============================
def load_or_extract_data():
    if os.path.exists(PROCESSED_DATA_FILE):
        print(f"Loading processed data from {PROCESSED_DATA_FILE} ...")
        data = np.load(PROCESSED_DATA_FILE)
        return data['X_img'], data['X_spec'], data['X_txt'], data['y']
    
    print("Extracting data from videos (first time)...")
    X_img, X_spec, X_txt, y = [],[],[],[]
    for folder in os.listdir(BASE_DATASET):
        path = os.path.join(BASE_DATASET, folder)
        if not os.path.isdir(path): continue
        label = 0 if folder.lower()=="aman" else 1
        for f in os.listdir(path):
            if not f.endswith(".mp4"): continue
            full = os.path.join(path, f)
            try:
                frame = extract_frame(full)
                spec, text_audio = extract_audio(full)
                text_ocr = extract_text_ocr(frame)
                txt_seq = text_to_sequence(text_audio + " " + text_ocr)
                X_img.append(frame)
                X_spec.append(spec)
                X_txt.append(txt_seq)
                y.append(label)
            except Exception as e:
                print("Skip:", f, e)
                continue
    X_img, X_spec, X_txt, y = np.array(X_img), np.array(X_spec), np.array(X_txt), np.array(y)
    np.savez(PROCESSED_DATA_FILE, X_img=X_img, X_spec=X_spec, X_txt=X_txt, y=y)
    print(f"Saved processed data to {PROCESSED_DATA_FILE}")
    return X_img, X_spec, X_txt, y

# ===============================
# ADAPTIVE FUSION
# ===============================
def adaptive_fusion(v,a,t):
    combined = layers.concatenate([v,a,t])
    weights = layers.Dense(3, activation='softmax')(combined)

    # Tidak menggunakan Lambda, aman untuk H5
    wv = layers.Reshape((1,))(layers.Lambda(lambda x: x[:,0])(weights))
    wa = layers.Reshape((1,))(layers.Lambda(lambda x: x[:,1])(weights))
    wt = layers.Reshape((1,))(layers.Lambda(lambda x: x[:,2])(weights))

    v = layers.Multiply()([v,wv])
    a = layers.Multiply()([a,wa])
    t = layers.Multiply()([t,wt])
    return layers.concatenate([v,a,t])

# ===============================
# BUILD MODEL
# ===============================
def build_model():
    img_in = layers.Input(shape=(*IMG_SIZE,3))
    spec_in = layers.Input(shape=(*SPEC_SIZE,1))
    txt_in = layers.Input(shape=(MAX_TEXT_LEN,))

    # VISUAL
    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    base.trainable = False
    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(128,activation='relu')(v)
    v = layers.Dropout(0.3)(v)

    # AUDIO
    a = layers.Conv2D(32,3,activation='relu',padding='same')(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64,3,activation='relu',padding='same')(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((1024,64))(a)
    a = layers.Bidirectional(layers.GRU(64, dropout=0.3))(a)
    a = layers.Dense(128,activation='relu')(a)

    # TEXT
    t = layers.Embedding(5000,128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64, dropout=0.3))(t)
    t = layers.Dense(128,activation='relu')(t)

    # FUSION
    fused = adaptive_fusion(v,a,t)

    # CLASSIFIER
    x = layers.Dense(256,activation='relu')(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128,activation='relu')(x)
    out = layers.Dense(1,activation='sigmoid')(x)

    model = models.Model([img_in,spec_in,txt_in], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ===============================
# VISUALIZATION
# ===============================
def plot_training_history(history):
    plt.figure(figsize=(12,5))
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.title('Accuracy')
    plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend()
    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.title('Loss')
    plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend()
    plt.show()

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,5))
    import seaborn as sns
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Aman','Tidak Aman'], yticklabels=['Aman','Tidak Aman'])
    plt.xlabel('Predicted'); plt.ylabel('True'); plt.title('Confusion Matrix')
    plt.show()

# ===============================
# MAIN
# ===============================
def main():
    X_img, X_spec, X_txt, y = load_or_extract_data()
    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_txt_tr, X_txt_val, y_tr, y_val = train_test_split(
        X_img, X_spec, X_txt, y, test_size=0.2, random_state=42, stratify=y
    )

    if os.path.exists(MODEL_FILE):
        print(f"Loading model from {MODEL_FILE} ...")
        model = tf.keras.models.load_model(MODEL_FILE)
        history = None
    else:
        model = build_model()
        es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        mc = tf.keras.callbacks.ModelCheckpoint(MODEL_FILE, monitor='val_loss', save_best_only=True)
        history = model.fit([X_img_tr,X_spec_tr,X_txt_tr], y_tr,
                            validation_data=([X_img_val,X_spec_val,X_txt_val], y_val),
                            epochs=EPOCHS, batch_size=BATCH_SIZE, callbacks=[es, mc])
        np.savez(HISTORY_FILE, history=history.history)

    # Evaluasi
    preds = model.predict([X_img_val,X_spec_val,X_txt_val])
    preds_class = (preds>0.5).astype(int)
    print(classification_report(y_val, preds_class))

    if history is not None:
        plot_training_history(history)
    plot_confusion_matrix(y_val, preds_class)

if __name__=="__main__":
    main()
```
