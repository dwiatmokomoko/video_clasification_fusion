```python
# -*- coding: utf-8 -*-
import os, cv2, re, numpy as np, warnings
import tensorflow as tf, librosa, whisper, easyocr, matplotlib.pyplot as plt
from moviepy import VideoFileClip
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
import itertools

warnings.filterwarnings("ignore")

# ===============================
# CONFIG
# ===============================
BASE_DATASET = r"D:\web\cnn_dnn_rnn_gan\Dataset"
IMG_SIZE = (224,224)
SPEC_SIZE = (128,128)
MAX_TEXT_LEN = 100
SR = 22050
DURATION = 3
BATCH_SIZE = 8
EPOCHS = 15

# ===============================
# LOAD MODELS (OCR + STT)
# ===============================
stt_model = whisper.load_model("base")
ocr_reader = easyocr.Reader(['id','en'], gpu=False)

# ===============================
# FEATURE EXTRACTION
# ===============================
def extract_frame(video_path):
    cap = cv2.VideoCapture(video_path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.set(cv2.CAP_PROP_POS_FRAMES, total//2)
    ret, frame = cap.read()
    cap.release()
    if not ret: return np.zeros((*IMG_SIZE,3))
    frame = cv2.resize(frame, IMG_SIZE)
    return (frame/255.0).astype(np.float32)

def extract_audio(video_path):
    tmp = "temp.wav"
    try:
        with VideoFileClip(video_path) as video:
            if video.audio is None: return np.zeros((*SPEC_SIZE,1)), ""
            video.audio.write_audiofile(tmp, logger=None)
        y, sr = librosa.load(tmp, sr=SR)
        if len(y) < SR*DURATION: y = np.pad(y, (0, SR*DURATION-len(y)))
        y = y[:SR*DURATION] + 0.005*np.random.randn(SR*DURATION)
        mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)
        mel = cv2.resize(librosa.power_to_db(mel), SPEC_SIZE)
        mel = np.expand_dims(mel, -1)
        mel = (mel - np.mean(mel))/(np.std(mel)+1e-9)
        text = ""
        try: text = stt_model.transcribe(tmp)['text']
        except: pass
        os.remove(tmp)
        return mel.astype(np.float32), text.lower()
    except:
        return np.zeros((*SPEC_SIZE,1)), ""

def extract_text_ocr(frame):
    try:
        frame_uint8 = (frame*255).astype(np.uint8)
        text = " ".join(ocr_reader.readtext(frame_uint8, detail=0))
        return text.lower()
    except: return ""

def text_to_sequence(text):
    text = re.sub(r'[^a-zA-Z ]', '', text)
    seq = [ord(c)%5000 for c in text][:MAX_TEXT_LEN]
    seq += [0]*(MAX_TEXT_LEN-len(seq))
    return np.array(seq)

# ===============================
# LOAD DATA
# ===============================
def load_data():
    imgs, specs, texts, labels = [],[],[],[]
    for folder in os.listdir(BASE_DATASET):
        path = os.path.join(BASE_DATASET, folder)
        if not os.path.isdir(path): continue
        label = 0 if folder.lower()=="aman" else 1
        for f in os.listdir(path):
            if not f.endswith(".mp4"): continue
            full = os.path.join(path, f)
            try:
                frame = extract_frame(full)
                spec, text_audio = extract_audio(full)
                text_ocr = extract_text_ocr(frame)
                full_text = text_audio + " " + text_ocr
                txt_seq = text_to_sequence(full_text)
                imgs.append(frame); specs.append(spec); texts.append(txt_seq); labels.append(label)
            except Exception as e:
                print("Skip:", f, e); continue
    print("Total data:", len(labels))
    return np.array(imgs), np.array(specs), np.array(texts), np.array(labels)

# ===============================
# MULTIMODAL GENERATOR
# ===============================
def multimodal_generator(X_img, X_spec, X_txt, y, batch_size=BATCH_SIZE):
    n = len(y)
    idx = np.arange(n)
    while True:
        np.random.shuffle(idx)
        for i in range(0, n, batch_size):
            batch_idx = idx[i:i+batch_size]
            yield [X_img[batch_idx], X_spec[batch_idx], X_txt[batch_idx]], y[batch_idx]

# ===============================
# ADAPTIVE FUSION
# ===============================
def adaptive_fusion(v,a,t):
    combined = layers.concatenate([v,a,t])
    weights = layers.Dense(3, activation='softmax')(combined)
    wv = layers.Lambda(lambda x: x[:,0:1])(weights)
    wa = layers.Lambda(lambda x: x[:,1:2])(weights)
    wt = layers.Lambda(lambda x: x[:,2:3])(weights)
    v = layers.Multiply()([v,wv])
    a = layers.Multiply()([a,wa])
    t = layers.Multiply()([t,wt])
    return layers.concatenate([v,a,t])

# ===============================
# BUILD MODEL
# ===============================
def build_model(fine_tune=False):
    # Inputs
    img_in = layers.Input(shape=(224,224,3))
    spec_in = layers.Input(shape=(128,128,1))
    txt_in = layers.Input(shape=(MAX_TEXT_LEN,))

    # VISUAL
    base = EfficientNetB0(include_top=False, weights='imagenet', input_tensor=img_in)
    if not fine_tune:
        base.trainable = False
    else:
        base.trainable = True
        for layer in base.layers[:-50]: layer.trainable=False
    v = layers.GlobalAveragePooling2D()(base.output)
    v = layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(1e-4))(v)
    v = layers.Dropout(0.3)(v)

    # AUDIO
    a = layers.Conv2D(32,3,activation='relu',padding='same', kernel_regularizer=regularizers.l2(1e-4))(spec_in)
    a = layers.MaxPooling2D()(a)
    a = layers.Conv2D(64,3,activation='relu',padding='same', kernel_regularizer=regularizers.l2(1e-4))(a)
    a = layers.MaxPooling2D()(a)
    a = layers.Reshape((1024,64))(a)
    a = layers.Bidirectional(layers.GRU(64, dropout=0.3, recurrent_dropout=0.3))(a)
    a = layers.Dense(128,activation='relu', kernel_regularizer=regularizers.l2(1e-4))(a)

    # TEXT
    t = layers.Embedding(5000,128)(txt_in)
    t = layers.Bidirectional(layers.LSTM(64, dropout=0.3, recurrent_dropout=0.3))(t)
    t = layers.Dense(128,activation='relu', kernel_regularizer=regularizers.l2(1e-4))(t)

    # FUSION
    fused = adaptive_fusion(v,a,t)

    # CLASSIFIER
    x = layers.Dense(256,activation='relu', kernel_regularizer=regularizers.l2(1e-4))(fused)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)
    x = layers.Dense(128,activation='relu', kernel_regularizer=regularizers.l2(1e-4))(x)
    out = layers.Dense(1, activation='sigmoid')(x)

    model = models.Model([img_in,spec_in,txt_in], out)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
    return model

# ===============================
# TRAIN MODEL USING tf.data.Dataset
# ===============================
def train_model(model, X_img_tr, X_spec_tr, X_txt_tr, y_tr,
                X_img_val, X_spec_val, X_txt_val, y_val):
    
    # Hitung class_weight
    class_weight = {0: y_tr.size/(2*np.sum(y_tr==0)),
                    1: y_tr.size/(2*np.sum(y_tr==1))}

    # Buat tf.data.Dataset untuk training
    train_dataset = tf.data.Dataset.from_tensor_slices(
        ((X_img_tr, X_spec_tr, X_txt_tr), y_tr)
    ).shuffle(buffer_size=1024).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    # Dataset validation
    val_dataset = tf.data.Dataset.from_tensor_slices(
        ((X_img_val, X_spec_val, X_txt_val), y_val)
    ).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    # Callbacks
    es = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
    mc = tf.keras.callbacks.ModelCheckpoint('best_model.h5', monitor='val_loss', save_best_only=True)

    # Fit model
    history = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=EPOCHS,
        class_weight=class_weight,
        callbacks=[es, mc]
    )

    # Prediksi dan laporan
    preds = model.predict(val_dataset)
    print("\nClassification Report:")
    print(classification_report(y_val, (preds>0.5).astype(int)))

    return model, history


# ===============================
# MAIN FUNCTION
# ===============================
def main():
    # Load dataset
    X_img, X_spec, X_txt, y = load_data()

    # Split train-validation
    X_img_tr, X_img_val, X_spec_tr, X_spec_val, X_txt_tr, X_txt_val, y_tr, y_val = train_test_split(
        X_img, X_spec, X_txt, y, test_size=0.2, stratify=y, random_state=42
    )

    # Build model
    model = build_model(fine_tune=True)
    model.summary()

    # Train
    model, history = train_model(
        model,
        X_img_tr, X_spec_tr, X_txt_tr, y_tr,
        X_img_val, X_spec_val, X_txt_val, y_val
    )

if __name__=="__main__":
    main()
```
